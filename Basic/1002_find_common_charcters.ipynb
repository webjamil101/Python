**1002. Find Common Characters**

**Easy**

**Companies** :

Given a string array words, return an array of all characters that show up in all strings within the words (including duplicates). You may return the answer in any order.

 

**Example 1:**
```python
Input: words = ["bella","label","roller"]
Output: ["e","l","l"]
```
**Example 2:**
```python
Input: words = ["cool","lock","cook"]
Output: ["c","o"]
```

**Constraints:**

- 1 <= words.length <= 100
- 1 <= words[i].length <= 100
- words[i] consists of lowercase English letters.

````python
from typing import List

class Solution:
    def commonChars(self, words: List[str]) -> List[str]:
        # Initialize a list to track the frequency of each letter (a-z)
        common = [float('inf')] * 26  # Use inf because we need to find the minimum frequency
        
        # Iterate through all words
        for word in words:
            # Create a frequency array for the current word
            temp = [0] * 26
            for char in word:
                temp[ord(char) - ord('a')] += 1
            
            # Update the common array to store the minimum frequency for each character
            for i in range(26):
                common[i] = min(common[i], temp[i])
        
        # Prepare the result list based on the common frequencies
        result = []
        for i in range(26):
            result.extend([chr(i + ord('a'))] * common[i])  # Add the character based on its frequency
        
        return result
````

````python
from typing import List
from collections import Counter

class Solution:
    def commonChars(self, words: List[str]) -> List[str]:
        # Initialize the result with the character counts of the first word
        common_count = Counter(words[0])
        
        # Iterate through each word and update the common character counts
        for word in words[1:]:
            # Get character counts for the current word
            word_count = Counter(word)
            
            # Keep only the minimum count of characters that appear in both `common_count` and `word_count`
            common_count &= word_count
        
        # Expand the result list based on the final common counts
        result = []
        for char, count in common_count.items():
            result.extend([char] * count)
        
        return result
````

````python
from typing import List

class Solution:
    def commonChars(self, words: List[str]) -> List[str]:
        # Sort each word to bring similar characters together
        words = [sorted(word) for word in words]
        
        # Start with the first word's characters
        result = []
        i = 0
        while i < len(words[0]):
            char = words[0][i]
            # Check if the character is present in the same position in all words
            if all(char in word for word in words):
                result.append(char)
                # Remove the first occurrence of the character from each word
                for j in range(len(words)):
                    words[j].remove(char)
            else:
                i += 1
        
        return result
````

````python
from typing import List

class Solution:
    def commonChars(self, words: List[str]) -> List[str]:
        # Initialize result list with characters of the first word
        result = list(words[0])
        
        # Iterate over each word after the first
        for word in words[1:]:
            temp_result = []
            # For each character in the current word, check if it exists in the result
            for char in result:
                if char in word:
                    temp_result.append(char)
                    word = word.replace(char, '', 1)  # Remove first occurrence of char from word
            result = temp_result  # Update result to be the common characters so far
        
        return result
````
