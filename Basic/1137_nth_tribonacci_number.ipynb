**1137. N-th Tribonacci Number**

**Easy**

**Companies**: **Coursera**

The Tribonacci sequence Tn is defined as follows: 

T0 = 0, T1 = 1, T2 = 1, and Tn+3 = Tn + Tn+1 + Tn+2 for n >= 0.

Given n, return the value of Tn.

 

**Example 1:**
```python
Input: n = 4
Output: 4
```
**Explanation:**
T_3 = 0 + 1 + 1 = 2
T_4 = 1 + 1 + 2 = 4

**Example 2:**
```python
Input: n = 25
Output: 1389537
 ```

**Constraints:**

- 0 <= n <= 37
- The answer is guaranteed to fit within a 32-bit integer, ie. answer <= 2^31 - 1.

````python
class Solution:
    def tribonacci(self, n: int) -> int:
        # Step 1: Create a memo dictionary to store computed T(n)
        memo = {0: 0, 1: 1, 2: 1}
        
        # Step 2: Recursive helper function
        def helper(k):
            if k in memo:
                return memo[k]
            # Step 3: Compute and memoize the result
            memo[k] = helper(k - 1) + helper(k - 2) + helper(k - 3)
            return memo[k]
        
        return helper(n)
````

````python
class Solution:
    def tribonacci(self, n: int) -> int:
        if n == 0: return 0
        if n == 1 or n == 2: return 1
        
        dp = [0] * (n + 1)
        dp[0], dp[1], dp[2] = 0, 1, 1
        
        for i in range(3, n + 1):
            dp[i] = dp[i - 1] + dp[i - 2] + dp[i - 3]
        
        return dp[n]
````

````python
class Solution:
    def tribonacci(self, n: int) -> int:
        if n == 0: return 0
        if n == 1 or n == 2: return 1

        a, b, c = 0, 1, 1  # T0, T1, T2
        for _ in range(3, n + 1):
            next_val = a + b + c
            a, b, c = b, c, next_val

        return c
````

````python
class Solution:
    ''' ✅ Efficient for very large n (not needed for n ≤ 37, but good to know)'''
    def tribonacci(self, n: int) -> int:
        def multiply(mat1, mat2):
            res = [[0] * 3 for _ in range(3)]
            for i in range(3):
                for j in range(3):
                    for k in range(3):
                        res[i][j] += mat1[i][k] * mat2[k][j]
            return res

        def matrix_power(mat, power):
            result = [[1 if i == j else 0 for j in range(3)] for i in range(3)]  # Identity
            while power:
                if power % 2 == 1:
                    result = multiply(result, mat)
                mat = multiply(mat, mat)
                power //= 2
            return result

        if n == 0: return 0
        if n == 1 or n == 2: return 1

        # Transformation matrix
        T = [
            [1, 1, 1],
            [1, 0, 0],
            [0, 1, 0]
        ]

        # Compute T^(n-2)
        T_n = matrix_power(T, n - 2)

        # Return T(n) = T_n[0][0]*1 + T_n[0][1]*1 + T_n[0][2]*0 = T_n[0][0] + T_n[0][1]
        return T_n[0][0] + T_n[0][1]
````
