# 💡 Problem Statement

> Given a 0-indexed integer array `nums` and an integer `k`, return the **number of pairs** `(i, j)` where:

- `0 <= i < j < n`
- `nums[i] == nums[j]`
- `(i * j) % k == 0`

---

## 🧪 Examples

### Example 1

```python
Input: nums = [3,1,2,2,2,1,3], k = 2
Output: 4
```

### Example 2

```python
Input: nums = [1,2,3,4], k = 1
Output: 0
```

---

## ✅ Constraints

- `1 <= nums.length <= 100` (small input size)
- `1 <= nums[i], k <= 100`

---

## ✅ Approach 1: Brute Force (Double Loop)

### 🔧 Idea

- Check **every possible pair (i, j)**.
- If `nums[i] == nums[j]` and `(i * j) % k == 0`, count it.

### ⏱ Time Complexity: `O(n^2)`

Works fine since `n <= 100`.

### ✅ Code

```python
class Solution:
    def countPairs(self, nums, k):
        n = len(nums)
        count = 0
        for i in range(n):
            for j in range(i + 1, n):
                if nums[i] == nums[j] and (i * j) % k == 0:
                    count += 1
        return count
```

---

## ⚙️ Approach 2: Optimized with HashMap (Grouping Indices)

### 🔧 Idea

- Group all indices where `nums[i]` occurs using a dictionary.
- For each group, check all unique pairs `(i, j)` from the list of indices where `nums[i] == nums[j]`, and then check `(i * j) % k == 0`.

### ⏱ Time Complexity:

- Grouping: `O(n)`
- Pair checking within groups: worst-case `O(n^2)` (still acceptable for small `n`)

### ✅ Code

```python
from collections import defaultdict

class Solution:
    def countPairs(self, nums, k):
        index_map = defaultdict(list)
        count = 0

        # Step 1: Group indices by number
        for i, num in enumerate(nums):
            index_map[num].append(i)

        # Step 2: For each group, check all valid pairs
        for indices in index_map.values():
            for i in range(len(indices)):
                for j in range(i + 1, len(indices)):
                    if (indices[i] * indices[j]) % k == 0:
                        count += 1
        return count
```

# **Count Equal and Divisible Pairs in an Array**

## **Problem Explanation**

Given an array `nums` and an integer `k`, we need to count all pairs of indices `(i, j)` where:

1. `i < j` (indices are ordered)
2. `nums[i] == nums[j]` (elements at these indices are equal)
3. `(i * j) % k == 0` (the product of indices is divisible by `k`)

### **Example 1**

**Input:** `nums = [3,1,2,2,2,1,3]`, `k = 2`  
**Output:** `4`  
**Explanation:**  
Valid pairs are:

- `(0, 6)` → `nums[0] == nums[6]` and `0 * 6 = 0` (divisible by `2`)
- `(2, 3)` → `nums[2] == nums[3]` and `2 * 3 = 6` (divisible by `2`)
- `(2, 4)` → `nums[2] == nums[4]` and `2 * 4 = 8` (divisible by `2`)
- `(3, 4)` → `nums[3] == nums[4]` and `3 * 4 = 12` (divisible by `2`)

### **Example 2**

**Input:** `nums = [1,2,3,4]`, `k = 1`  
**Output:** `0`  
**Explanation:** No duplicates exist, so no valid pairs.

---

## **Approach**

1. **Brute-Force Check**:

   - Iterate through all possible pairs `(i, j)` where `i < j`.
   - For each pair, check if:
     - `nums[i] == nums[j]`
     - `(i * j) % k == 0`
   - Count all such valid pairs.

2. **Optimization Insight**:
   - Since the constraints are small (`nums.length <= 100`), a brute-force approach is efficient.
   - No need for complex data structures.

---

## **Solution Code**

```python
def countPairs(nums, k):
    count = 0
    n = len(nums)
    for i in range(n):
        for j in range(i + 1, n):
            if nums[i] == nums[j] and (i * j) % k == 0:
                count += 1
    return count
```

### **Explanation**

1. **Initialization**:

   - `count` keeps track of valid pairs.
   - `n` is the length of the array.

2. **Nested Loops**:

   - Outer loop (`i`) runs from `0` to `n-1`.
   - Inner loop (`j`) runs from `i+1` to `n-1` (ensuring `i < j`).

3. **Condition Check**:

   - If `nums[i] == nums[j]` and `(i * j) % k == 0`, increment `count`.

4. **Return Result**:
   - The total count of valid pairs is returned.

---

## **Complexity Analysis**

- **Time Complexity**: **O(n²)**

  - We check all possible pairs, leading to nested loops.
  - For `n = 100`, this results in **100 × 99 / 2 = 4950** iterations (acceptable).

- **Space Complexity**: **O(1)**
  - Only a few variables (`count`, `i`, `j`) are used.

---

## **Key Takeaways**

- **Brute-force is sufficient** due to small constraints (`n ≤ 100`).
- **Check all pairs** where `i < j` for the given conditions.
- **No extra space** is needed, making it memory efficient.

This approach efficiently solves the problem while being easy to understand and implement. 🚀

Instead of `defaultdict`, we'll use a regular dictionary and check if a key exists before appending to it.

## ✅ Final Code (No `defaultdict`)

```python
class Solution:
    def countPairs(self, nums, k):
        index_map = {}  # Use regular dict instead of defaultdict
        count = 0

        # Step 1: Group indices by number
        for i, num in enumerate(nums):
            if num not in index_map:
                index_map[num] = []
            index_map[num].append(i)

        # Step 2: For each group, check valid pairs
        for indices in index_map.values():
            for i in range(len(indices)):
                for j in range(i + 1, len(indices)):
                    if (indices[i] * indices[j]) % k == 0:
                        count += 1
        return count
```

## ✅ Problem

You're writing a class method `countGood` inside the `Solution` class, but **you forgot to include `self`** as the first parameter. In Python, all instance methods must include `self`.

---

## ❌ Your code (faulty part):

```python
class Solution:
    def countGood(nums, k):   # ❌ Missing 'self'
```

This will raise an error when called using an instance of `Solution`, or even statically in some cases depending on how it's invoked.

---

## ✅ Corrected Version:

```python
from collections import defaultdict

class Solution:
    def countGood(self, nums, k):  # ✅ 'self' added
        freq = defaultdict(int)
        n = len(nums)
        left = 0
        count = 0
        res = 0

        for right in range(n):
            count += freq[nums[right]]
            freq[nums[right]] += 1

            while count >= k:
                res += n - right
                freq[nums[left]] -= 1
                count -= freq[nums[left]]
                left += 1

        return res
```

## ✅ Optional: Static Method Version

If you don't want to use `self` (i.e., not part of an instance), declare it as a `@staticmethod`:

```python
from collections import defaultdict

class Solution:
    @staticmethod
    def countGood(nums, k):  # ✅ Now it's a static method
        freq = defaultdict(int)
        n = len(nums)
        left = 0
        count = 0
        res = 0

        for right in range(n):
            count += freq[nums[right]]
            freq[nums[right]] += 1

            while count >= k:
                res += n - right
                freq[nums[left]] -= 1
                count -= freq[nums[left]]
                left += 1

        return res
```
