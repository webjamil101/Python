# 263. Ugly Number

## Problem Statement

An **ugly number** is a positive integer that does not have a prime factor other than **2, 3, and 5**.

Given an integer `n`, return `true` if `n` is an ugly number.

## Example 1:

**Input:**

```
n = 6
```

**Output:**

```
true
```

**Explanation:**  
6 can be factored as `2 Ã— 3`, which only includes the prime factors `2`, `3`, and `5`.

---

## Example 2:

**Input:**

```
n = 1
```

**Output:**

```
true
```

**Explanation:**  
1 has no prime factors and is considered an ugly number.

---

## Example 3:

**Input:**

```
n = 14
```

**Output:**

```
false
```

**Explanation:**  
14 includes the prime factor `7`, which is not allowed.

---

## Constraints

- `-2^31 <= n <= 2^31 - 1`


````python
def isUgly(n: int) -> bool:
    if n <= 0:
        return False  # Ugly numbers must be positive

    # Keep dividing n by 2, 3, or 5 while possible
    for factor in [2, 3, 5]:
        while n % factor == 0:
            n //= factor

    return n == 1  # If n reduces to 1, it is an ugly number

# Test Cases
test_cases = [6, 1, 14, 30, 0, -6, 25, 8, 15, 21, 7, 2**31 - 1, -2**31, 45]

for num in test_cases:
    print(f"isUgly({num}) = {isUgly(num)}")
````
