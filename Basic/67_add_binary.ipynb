**67. Add Binary**

**Easy**

**Companies**: Adobe Amazon Apple Facebook Google Microsoft Oracle

Given two binary strings a and b, return their sum as a binary string.

**Example 1:**

```python
Input: a = "11", b = "1"
Output: "100"
```

**Example 2:**

```python
Input: a = "1010", b = "1011"
Output: "10101"
```

**Constraints:**

- 1 <= a.length, b.length <= 104
- a and b consist only of '0' or '1' characters.
- Each string does not contain leading zeros except for the zero itself.


````python
# =====================================================
# ðŸ§  Approach 1: Built-in Conversion
# -----------------------------------------------------
# ðŸ”¹ Algorithm:
#   1. Convert 'a' and 'b' to integers using int(a, 2) and int(b, 2).
#   2. Add them.
#   3. Convert result to binary with bin(sum)[2:].

# ðŸ”¹ Time: O(n)
# ðŸ”¹ Space: O(1)
# =====================================================
class Solution:
    def addBinary(self, a: str, b: str) -> str:
        return bin(int(a, 2) + int(b, 2))[2:]
````

````python
# =====================================================
# ðŸ§  Approach 2: Manual Binary Addition
# -----------------------------------------------------
# ðŸ”¹ Algorithm:
#   1. Use two pointers i, j at ends of a, b.
#   2. Initialize carry = 0.
#   3. While i or j or carry:
#       - Get bits ai, bj (0 if index < 0)
#       - total = ai + bj + carry
#       - append total % 2 to result
#       - carry = total // 2
#   4. Reverse result and join.
#
# ðŸ”¹ Time: O(max(len(a), len(b)))
# ðŸ”¹ Space: O(max(len(a), len(b)))
# =====================================================
class Solution:
    def addBinary(self, a: str, b: str) -> str:
        i, j = len(a) - 1, len(b) - 1
        carry = 0
        res = []

        while i >= 0 or j >= 0 or carry:
            total = carry
            if i >= 0:
                total += int(a[i])
                i -= 1
            if j >= 0:
                total += int(b[j])
                j -= 1

            res.append(str(total % 2))
            carry = total // 2

        return ''.join(reversed(res))
````

````python
# =====================================================
# ðŸ§  Approach 3: Using zfill() for Alignment
# -----------------------------------------------------
# ðŸ”¹ Algorithm:
#   1. Pad both strings to same length: a, b = a.zfill(n), b.zfill(n)
#   2. Traverse from right to left.
#   3. For each bit, sum with carry â†’ record result and update carry.
#
# ðŸ”¹ Time: O(n)
# ðŸ”¹ Space: O(n)
# =====================================================
class Solution:
    def addBinary(self, a: str, b: str) -> str:
        n = max(len(a), len(b))
        a, b = a.zfill(n), b.zfill(n)
        carry = 0
        res = []

        for i in range(n - 1, -1, -1):
            total = int(a[i]) + int(b[i]) + carry
            res.append(str(total % 2))
            carry = total // 2

        if carry:
            res.append('1')

        return ''.join(reversed(res))
````

````python
# =====================================================
# ðŸ§  Approach 4: Bit Manipulation (Full Adder Logic)
# -----------------------------------------------------
# ðŸ”¹ Algorithm:
#   1. Convert binary strings to integers x, y.
#   2. While y != 0:
#        - answer = x ^ y
#        - carry = (x & y) << 1
#        - x = answer
#        - y = carry
#   3. Convert x to binary string.
#
# ðŸ”¹ Time: O(1) average, O(n) worst (limited by int size)
# ðŸ”¹ Space: O(1)
# =====================================================
class Solution:
    def addBinary(self, a: str, b: str) -> str:
        x, y = int(a, 2), int(b, 2)
        while y:
            x, y = x ^ y, (x & y) << 1
        return bin(x)[2:]
````

````python
# =====================================================
# ðŸ§  Approach 5: Using deque
# -----------------------------------------------------
# ðŸ”¹ Same logic as manual addition, but uses deque to
#   avoid reversing the list later.
#
# ðŸ”¹ Time: O(n)
# ðŸ”¹ Space: O(n)
# =====================================================
from collections import deque
class Solution:
    def addBinary(self, a: str, b: str) -> str:
        i, j = len(a) - 1, len(b) - 1
        carry = 0
        res = deque()

        while i >= 0 or j >= 0 or carry:
            total = carry
            if i >= 0:
                total += int(a[i])
                i -= 1
            if j >= 0:
                total += int(b[j])
                j -= 1
            res.appendleft(str(total % 2))
            carry = total // 2

        return ''.join(res)
````
