**1207. Unique Number of Occurrences**

**Easy**

**Company: Google**

Given an array of integers arr, return true if the number of occurrences of each value in the array is unique or false otherwise.

**Example 1:**

```python
Input: arr = [1,2,2,1,1,3]
Output: true
```

**Explanation:** The value 1 has 3 occurrences, 2 has 2 and 3 has 1. No two values have the same number of occurrences.

**Example 2:**

```python
Input: arr = [1,2]
Output: false
```

**Example 3:**

```python
Input: arr = [-3,0,1,-3,1,1,1,-3,10,0]
Output: true
```

**Constraints:**

- 1 <= arr.length <= 1000
- -1000 <= arr[i] <= 1000


````python
import collections

class Solution:
    def uniqueOccurrences(self, arr: list[int]) -> bool:
        # Algorithm:
        # 1. Use a hash map (dictionary) to count the occurrences of each number.
        # 2. Iterate through the input array and populate the map.
        # 3. Use a set to store the frequencies.
        # 4. Iterate through the values (frequencies) of the hash map.
        # 5. For each frequency, try to add it to the set. If the set
        #    already contains the frequency, it means the occurrences are not unique,
        #    so we return False.
        # 6. If the loop completes without finding duplicate frequencies, return True.

        # Step 1 & 2: Count frequencies of numbers in the array.
        frequency_map = collections.Counter(arr)
        
        # Step 3 & 4: Store and check for unique frequencies.
        unique_frequencies = set()
        for freq in frequency_map.values():
            if freq in unique_frequencies:
                return False
            unique_frequencies.add(freq)
            
        return True
````

````python
class Solution:
    def uniqueOccurrences(self, arr: list[int]) -> bool:
        # Algorithm:
        # 1. Use a fixed-size array (of size 2001) to act as a frequency map.
        #    We offset the values by 1000 to handle negative numbers.
        # 2. Iterate through the input array `arr` and populate this frequency array.
        # 3. Sort the frequency array.
        # 4. Iterate through the sorted array. If any two consecutive non-zero elements
        #    are the same, it means we have duplicate frequencies, so return False.
        # 5. If no duplicate frequencies are found, return True.

        # The range of numbers is [-1000, 1000], so an array of size 2001 is sufficient.
        freq_array = [0] * 2001
        
        # Count frequencies, using an offset to handle negative numbers.
        for num in arr:
            freq_array[num + 1000] += 1
            
        # Sort the frequency array to easily check for duplicates.
        freq_array.sort()
        
        # Iterate through the sorted frequencies.
        for i in range(1, 2001):
            # Check for duplicate non-zero frequencies.
            if freq_array[i] != 0 and freq_array[i] == freq_array[i-1]:
                return False
                
        return True
````
