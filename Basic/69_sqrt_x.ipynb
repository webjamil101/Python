**69. Sqrt(x)**

**Easy**

**Companies**: Amazon Apple Bloomberg Facebook Google LinkedIn Lyft Microsoft Qualtrics Uber Wayfair

Given a non-negative integer x, return the square root of x rounded down to the nearest integer. The returned integer should be non-negative as well.

You must not use any built-in exponent function or operator.

- For example, do not use pow(x, 0.5) in c++ or x \*\* 0.5 in python.

**Example 1:**

```python
Input: x = 4
Output: 2
```

**Explanation:** The square root of 4 is 2, so we return 2.

**Example 2:**

```python
Input: x = 8
Output: 2
```

**Explanation:** The square root of 8 is 2.82842..., and since we round it down to the nearest integer, 2 is returned.

**Constraints:**

- 0 <= x <= 231 - 1


````python
# =====================================================
# ðŸ§  Approach 1: Brute Force Linear Search
# -----------------------------------------------------
# ðŸ”¹ Algorithm:
#   1. Handle small values: if x < 2 â†’ return x.
#   2. Loop from 1 to x:
#        - If i*i == x â†’ return i.
#        - If i*i > x â†’ return i - 1.
#
# ðŸ”¹ Time Complexity: O(âˆšx)
# ðŸ”¹ Space Complexity: O(1)
# =====================================================
class Solution:
    def mySqrt(self, x: int) -> int:
        if x < 2:
            return x

        i = 1
        while i * i <= x:
            i += 1
        return i - 1
````

````python
# =====================================================
# ðŸ§  Approach 2: Binary Search
# -----------------------------------------------------
# ðŸ”¹ Algorithm:
#   1. Handle x < 2 case directly.
#   2. Set left=1, right=x//2.
#   3. While left <= right:
#        - mid = (left + right)//2
#        - if mid*mid == x: return mid
#        - if mid*mid < x: left = mid + 1
#        - else: right = mid - 1
#   4. Return right (largest mid where mid*mid <= x)
#
# ðŸ”¹ Time Complexity: O(log x)
# ðŸ”¹ Space Complexity: O(1)
# =====================================================
class Solution:
    def mySqrt(self, x: int) -> int:
        if x < 2:
            return x

        left, right = 1, x // 2

        while left <= right:
            mid = (left + right) // 2
            sq = mid * mid

            if sq == x:
                return mid
            elif sq < x:
                left = mid + 1
            else:
                right = mid - 1

        return right
````

````python
# =====================================================
# ðŸ§  Approach 3: Newtonâ€™s Method (Iterative)
# -----------------------------------------------------
# ðŸ”¹ Formula:
#   next_guess = (guess + x / guess) / 2
#
# ðŸ”¹ Algorithm:
#   1. If x < 2 â†’ return x.
#   2. guess = x / 2
#   3. While guess*guess > x:
#        guess = (guess + x / guess) / 2
#   4. Return int(guess)
#
# ðŸ”¹ Time Complexity: O(log x)
# ðŸ”¹ Space Complexity: O(1)
# =====================================================
class Solution:
    def mySqrt(self, x: int) -> int:
        if x < 2:
            return x

        guess = x / 2
        while guess * guess > x:
            guess = (guess + x / guess) / 2

        return int(guess)
````

````python
# =====================================================
# ðŸ§  Approach 4: Bit Manipulation (Binary Root)
# -----------------------------------------------------
# ðŸ”¹ Algorithm:
#   1. Initialize res = 0
#   2. For bit in [1 << 15, 1 << 14, ..., 1]:
#       - if (res + bit)^2 <= x: res += bit
#   3. Return res
#
# ðŸ”¹ Time Complexity: O(log x)
# ðŸ”¹ Space Complexity: O(1)
# =====================================================
class Solution:
    def mySqrt(self, x: int) -> int:
        res = 0
        bit = 1 << 15  # 2^15 â‰ˆ 32768, enough for sqrt(2^31)

        while bit > 0:
            if (res + bit) * (res + bit) <= x:
                res += bit
            bit >>= 1

        return res
````

````python
# =====================================================
# ðŸ§  Approach 5: Using math.isqrt() (Python 3.8+)
# -----------------------------------------------------
# ðŸ”¹ Direct built-in integer square root (no float math)
# ðŸ”¹ Time Complexity: O(1)
# =====================================================
import math
class Solution:
    def mySqrt(self, x: int) -> int:
        return math.isqrt(x)
````
