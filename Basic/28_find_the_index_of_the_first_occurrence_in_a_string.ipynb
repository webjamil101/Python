**28. Find the Index of the First Occurrence in a String**

**Easy**

**Companies**: Amazon Apple Bloomberg Citadel Dropbox eBay Facebook Google Microsoft Oracle Pocket Gems VMware

Given two strings needle and haystack, return the index of the first occurrence of needle in haystack, or -1 if needle is not part of haystack.

**Example 1:**

```python
Input: haystack = "sadbutsad", needle = "sad"
Output: 0
```

**Explanation:** "sad" occurs at index 0 and 6.

The first occurrence is at index 0, so we return 0.

**Example 2:**

```python
Input: haystack = "leetcode", needle = "leeto"
Output: -1
```

**Explanation:** "leeto" did not occur in "leetcode", so we return -1.

**Constraints:**

- 1 <= haystack.length, needle.length <= 104
- haystack and needle consist of only lowercase English characters.


````python
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        """
        Approach 1: Python Built-in find()
        ----------------------------------
        Algorithm:
        1. Use built-in string find() which returns:
              - index of first match
              - -1 if no match
        2. Works very efficiently and uses optimized C code.

        Time Complexity:  O(n) average
        Space Complexity: O(1)
        """
        return haystack.find(needle)
````

````python
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        """
        Approach 2: Sliding Window Brute Force
        --------------------------------------
        Algorithm:
        1. Let n = len(haystack), m = len(needle)
        2. Loop i from 0 to n - m:
              - Compare haystack[i : i+m] with needle
              - If match → return i
        3. If no match found → return -1

        Time Complexity:  O(n*m)
        Space Complexity: O(1)
        """
        n, m = len(haystack), len(needle)

        for i in range(n - m + 1):
            if haystack[i:i+m] == needle:
                return i

        return -1
````

````python
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        """
        Approach 3: Two-Pointer Manual Match
        -------------------------------------
        Algorithm:
        1. Let i scan haystack.
        2. For each i, try to match all chars of needle:
              - j scans needle
              - If mismatch → break
              - If j reaches end → found, return i
        3. If no match → return -1

        Time:  O(n*m)
        Space: O(1)
        """
        n, m = len(haystack), len(needle)

        for i in range(n - m + 1):
            match = True
            for j in range(m):
                if haystack[i+j] != needle[j]:
                    match = False
                    break
            if match:
                return i

        return -1
````
