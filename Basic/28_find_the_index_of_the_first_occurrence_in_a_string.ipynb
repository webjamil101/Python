## 28. Find the Index of the First Occurrence in a String

**Difficulty:** Easy  
**Topics:** String Manipulation

### Problem Statement:

Given two strings `needle` and `haystack`, return the index of the **first occurrence** of `needle` in `haystack`, or `-1` if `needle` is not part of `haystack`.

### Example 1:

**Input:**

```text
haystack = "sadbutsad", needle = "sad"
```

**Output:**

```text
0
```

**Explanation:**  
"sad" occurs at index 0 and 6. The first occurrence is at index 0, so we return 0.

### Example 2:

**Input:**

```text
haystack = "leetcode", needle = "leeto"
```

**Output:**

```text
-1
```

**Explanation:**  
"leeto" did not occur in "leetcode", so we return -1.

### Constraints:

- `1 <= haystack.length, needle.length <= 10^4`
- `haystack` and `needle` consist of only lowercase English characters.

---

### Solution:

You can solve this problem using Python's built-in method `find()` which returns the index of the first occurrence of `needle` in `haystack`.

#### Python Code:

```python
def strStr(haystack: str, needle: str) -> int:
    return haystack.find(needle)
```

### Explanation:

- The `find()` method searches for the substring `needle` in the string `haystack`.
- If `needle` is found, `find()` returns the index of its first occurrence.
- If `needle` is not found, `find()` returns `-1`.

#### Example Usage:

```python
print(strStr("sadbutsad", "sad"))  # Output: 0
print(strStr("leetcode", "leeto")) # Output: -1
```


`````python
def strStr(haystack: str, needle: str) -> int:
    # Return the index of the first occurrence of `needle` in `haystack`
    return haystack.find(needle)

# Test Cases

# Test Case 1: Basic test with the needle present
haystack = "sadbutsad"
needle = "sad"
print(strStr(haystack, needle))  # Expected Output: 0
# Explanation: "sad" occurs at index 0 and 6. The first occurrence is at index 0.

# Test Case 2: Test with the needle not present in the haystack
haystack = "leetcode"
needle = "leeto"
print(strStr(haystack, needle))  # Expected Output: -1
# Explanation: "leeto" did not occur in "leetcode", so return -1.

# Test Case 3: Test with needle equal to haystack
haystack = "hello"
needle = "hello"
print(strStr(haystack, needle))  # Expected Output: 0
# Explanation: The needle is the same as the haystack, so the first occurrence is at index 0.

# Test Case 4: Test with an empty needle
haystack = "hello"
needle = ""
print(strStr(haystack, needle))  # Expected Output: 0
# Explanation: An empty needle is found at the start of any string, so return 0.

# Test Case 5: Test with an empty haystack
haystack = ""
needle = "abc"
print(strStr(haystack, needle))  # Expected Output: -1
# Explanation: An empty haystack cannot contain any needle, so return -1.

# Test Case 6: Test with needle longer than haystack
haystack = "short"
needle = "longerneedle"
print(strStr(haystack, needle))  # Expected Output: -1
# Explanation: The needle is longer than the haystack, so it's impossible to find it, return -1.

# Test Case 7: Case sensitivity test
haystack = "Hello, welcome to Python!"
needle = "hello"
print(strStr(haystack, needle))  # Expected Output: -1
# Explanation: "hello" and "Hello" are different because the search is case-sensitive.

# Test Case 8: Test with the first occurrence at a non-zero index
haystack = "this is a test sentence"
needle = "test"
print(strStr(haystack, needle))  # Expected Output: 10
# Explanation: "test" occurs at index 10 in the string.

# Test Case 9: Test with large inputs (performance check)
haystack = "a" * 10000  # Haystack of 10,000 'a' characters
needle = "a" * 5000  # Needle of 5000 'a' characters
print(strStr(haystack, needle))  # Expected Output: 0
# Explanation: The needle occurs at the very beginning of the haystack, so return 0.
`````

`````python
class StringFinder:
    def __init__(self, haystack: str):
        self.haystack = haystack

    def strStr(self, needle: str) -> int:
        """
        Returns the index of the first occurrence of `needle` in the `haystack`
        or -1 if `needle` is not part of `haystack`.
        """
        return self.haystack.find(needle)

# Test Cases
if __name__ == "__main__":
    # Test Case 1: Basic test with the needle present
    finder1 = StringFinder("sadbutsad")
    print(finder1.strStr("sad"))  # Expected Output: 0

    # Test Case 2: Test with the needle not present in the haystack
    finder2 = StringFinder("leetcode")
    print(finder2.strStr("leeto"))  # Expected Output: -1

    # Test Case 3: Test with needle equal to haystack
    finder3 = StringFinder("hello")
    print(finder3.strStr("hello"))  # Expected Output: 0

    # Test Case 4: Test with an empty needle
    finder4 = StringFinder("hello")
    print(finder4.strStr(""))  # Expected Output: 0

    # Test Case 5: Test with an empty haystack
    finder5 = StringFinder("")
    print(finder5.strStr("abc"))  # Expected Output: -1

    # Test Case 6: Test with needle longer than haystack
    finder6 = StringFinder("short")
    print(finder6.strStr("longerneedle"))  # Expected Output: -1

    # Test Case 7: Case sensitivity test
    finder7 = StringFinder("Hello, welcome to Python!")
    print(finder7.strStr("hello"))  # Expected Output: -1

    # Test Case 8: Test with the first occurrence at a non-zero index
    finder8 = StringFinder("this is a test sentence")
    print(finder8.strStr("test"))  # Expected Output: 10

    # Test Case 9: Test with large inputs (performance check)
    finder9 = StringFinder("a" * 10000)  # Haystack of 10,000 'a' characters
    print(finder9.strStr("a" * 5000))  # Expected Output: 0
`````
