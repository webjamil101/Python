# 2900. Longest Unequal Adjacent Groups Subsequence I

# Easy

You are given a string array words and a binary array groups both of length n, where words[i] is associated with groups[i].

Your task is to select the longest alternating subsequence from words. A subsequence of words is alternating if for any two consecutive strings in the sequence, their corresponding elements in the binary array groups differ. Essentially, you are to choose strings such that adjacent elements have non-matching corresponding bits in the groups array.

Formally, you need to find the longest subsequence of an array of indices [0, 1, ..., n - 1] denoted as [i0, i1, ..., ik-1], such that groups[ij] != groups[ij+1] for each 0 <= j < k - 1 and then find the words corresponding to these indices.

Return the selected subsequence. If there are multiple answers, return any of them.

> Note: The elements in words are distinct.

# Example 1:

```
Input: words = ["e","a","b"], groups = [0,0,1]

Output: ["e","b"]

Explanation: A subsequence that can be selected is ["e","b"] because groups[0] != groups[2]. Another subsequence that can be selected is ["a","b"] because groups[1] != groups[2]. It can be demonstrated that the length of the longest subsequence of indices that satisfies the condition is 2.
```

# Example 2:

```
Input: words = ["a","b","c","d"], groups = [1,0,1,1]

Output: ["a","b","c"]

Explanation: A subsequence that can be selected is ["a","b","c"] because groups[0] != groups[1] and groups[1] != groups[2]. Another subsequence that can be selected is ["a","b","d"] because groups[0] != groups[1] and groups[1] != groups[3]. It can be shown that the length of the longest subsequence of indices that satisfies the condition is 3.

```

- Constraints:

* 1 <= n == words.length == groups.length <= 100
* 1 <= words[i].length <= 10
* groups[i] is either 0 or 1.
* words consists of distinct strings.
* words[i] consists of lowercase English letters.


````python
from typing import List
import unittest

class Solution:
    def getLongestSubsequence(self, words: List[str], groups: List[int]) -> List[str]:
        """
        Calculates the longest valid subsequence of words based on alternating groups.

        A valid subsequence requires that adjacent words belong to different groups.
        If multiple longest valid subsequences exist, the one that comes first
        lexicographically is returned. Due to the problem's structure and constraints
        (N <= 100, words are ordered), a greedy approach of simply appending the
        next valid word achieves both maximum length and the desired lexicographical
        order among longest subsequences.

        Args:
            words: A 0-indexed string array of words.
            groups: A binary array where groups[i] is the group of words[i].

        Returns:
            A string array representing the words corresponding to a longest valid subsequence.
        """
        
        longest_subsequence_words = []

        # Constraints: 1 <= n <= 100, so words and groups will always have at least one element.
        # We can always start the subsequence with the first word.
        
        # Add the very first word and its group to initialize the subsequence
        longest_subsequence_words.append(words[0])
        last_group = groups[0]

        # Iterate through the rest of the words starting from the second one (index 1)
        for i in range(1, len(words)):
            current_word = words[i]
            current_group = groups[i]

            # If the current word's group is different from the group of the last word
            # added to our subsequence, we can extend the subsequence.
            if current_group != last_group:
                longest_subsequence_words.append(current_word)
                last_group = current_group # Update the last_group for the next comparison

        return longest_subsequence_words

# --- Test Cases ---
class TestLongestUnequalAdjacentGroupsSubsequence(unittest.TestCase):

    def setUp(self):
        self.solution = Solution()

    # Example 1 from problem description
    def test_example_1(self):
        words = ["e", "a", "b"]
        groups = [0, 0, 1]
        expected = ["e", "b"]
        self.assertEqual(self.solution.getLongestSubsequence(words, groups), expected)

    # Example 2 from problem description (not explicitly given, but derived)
    def test_example_2(self):
        words = ["a", "b", "c", "d"]
        groups = [1, 0, 1, 0]
        expected = ["a", "b", "c", "d"]
        self.assertEqual(self.solution.getLongestSubsequence(words, groups), expected)

    # Example from problem description (not explicitly given, but derived)
    def test_example_3(self):
        words = ["a", "b", "c", "d"]
        groups = [0, 0, 0, 0]
        expected = ["a"]
        self.assertEqual(self.solution.getLongestSubsequence(words, groups), expected)
        
    # Edge Case: Single word
    def test_single_word(self):
        words = ["hello"]
        groups = [0]
        expected = ["hello"]
        self.assertEqual(self.solution.getLongestSubsequence(words, groups), expected)

    # Edge Case: All same groups
    def test_all_same_groups(self):
        words = ["one", "two", "three", "four"]
        groups = [1, 1, 1, 1]
        expected = ["one"]
        self.assertEqual(self.solution.getLongestSubsequence(words, groups), expected)

    # Edge Case: Alternating groups from start
    def test_perfect_alternating(self):
        words = ["apple", "banana", "cherry", "date", "elderberry"]
        groups = [0, 1, 0, 1, 0]
        expected = ["apple", "banana", "cherry", "date", "elderberry"]
        self.assertEqual(self.solution.getLongestSubsequence(words, groups), expected)

    # Edge Case: Alternating groups, then a block of same groups
    def test_alternating_then_block(self):
        words = ["x", "y", "z", "a", "b"]
        groups = [0, 1, 0, 0, 0]
        expected = ["x", "y", "z"]
        self.assertEqual(self.solution.getLongestSubsequence(words, groups), expected)

    # Edge Case: Block of same groups, then alternating
    def test_block_then_alternating(self):
        words = ["x", "y", "z", "a", "b"]
        groups = [0, 0, 0, 1, 0]
        expected = ["x", "a", "b"]
        self.assertEqual(self.solution.getLongestSubsequence(words, groups), expected)

    # Test with mixed groups and multiple words with same group following a valid one
    def test_mixed_groups(self):
        words = ["hi", "there", "friend", "how", "are", "you"]
        groups = [0, 1, 1, 0, 1, 1]
        expected = ["hi", "there", "how", "are"] # 'friend' group 1, skip; 'you' group 1, skip
        self.assertEqual(self.solution.getLongestSubsequence(words, groups), expected)

    # Test with words having different lengths
    def test_different_word_lengths(self):
        words = ["a", "bb", "ccc", "dddd", "e"]
        groups = [0, 1, 0, 1, 0]
        expected = ["a", "bb", "ccc", "dddd", "e"]
        self.assertEqual(self.solution.getLongestSubsequence(words, groups), expected)

    # Test with lexicographical order - this algorithm inherently picks the "first"
    # because it iterates from left to right.
    def test_lexicographical_preference(self):
        # If we had "a", "b" and "c", "d" where groups could be same for "a" and "c"
        # but current solution must be longest.
        # This test case shows that if a word is valid, it's always taken.
        words = ["apple", "zebra", "banana", "yak"]
        groups = [0, 1, 0, 1]
        expected = ["apple", "zebra", "banana", "yak"]
        self.assertEqual(self.solution.getLongestSubsequence(words, groups), expected)

        words = ["zoo", "cat", "dog", "ant"]
        groups = [0, 1, 0, 1]
        expected = ["zoo", "cat", "dog", "ant"]
        self.assertEqual(self.solution.getLongestSubsequence(words, groups), expected)


if __name__ == '__main__':
    print("Running tests for LeetCode 2900: Longest Unequal Adjacent Groups Subsequence I")
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
````
