**26. Remove Duplicates from Sorted Array**

**Easy**

**Companies**:Google, Amazon,Microsoft,Linkedin,expedia

Given an integer array nums sorted in non-decreasing order, remove the duplicates in-place such that each unique element appears only once. The relative order of the elements should be kept the same. Then return the number of unique elements in nums.

Consider the number of unique elements of nums to be k, to get accepted, you need to do the following things:

- Change the array nums such that the first k elements of nums contain the unique elements in the order they were present in nums initially. The remaining elements of nums are not important as well as the size of nums.

- Return k.

**Custom Judge**:

The judge will test your solution with the following code:

- int[] nums = [...]; // Input array
- int[] expectedNums = [...]; // The expected answer with correct length

- int k = removeDuplicates(nums); // Calls your implementation

- assert k == expectedNums.length;
- for (int i = 0; i < k; i++) {
  assert nums[i] == expectedNums[i];
  }

If all assertions pass, then your solution will be accepted.

**Example 1**:

```python
Input: nums = [1,1,2]
Output: 2, nums = [1,2,_]
```

**Explanation**: Your function should return k = 2, with the first two elements of nums being 1 and 2 respectively.
It does not matter what you leave beyond the returned k (hence they are underscores).

**Example 2**:

```python
Input: nums = [0,0,1,1,1,2,2,3,3,4]
Output: 5, nums = [0,1,2,3,4,_,_,_,_,_]
```

**Explanation**: Your function should return k = 5, with the first five elements of nums being 0, 1, 2, 3, and 4 respectively.
It does not matter what you leave beyond the returned k (hence they are underscores).

**Constraints**:

- 1 <= nums.length <= 3 \* 104
- -100 <= nums[i] <= 100
- nums is sorted in non-decreasing order.


````python
# --- Approach 1: Two Pointers (Optimal) ---
# This is the most efficient and standard way to solve the problem
# due to the "in-place" and "sorted array" constraints.
#
# Algorithm:
# 1. Initialize a 'write' pointer `k` to 1, which will track the
#    position for the next unique element to be written. The first
#    element of the sorted array is always unique, so we start from index 1.
# 2. Use a 'read' pointer `i` to iterate through the array from the second
#    element (index 1) to the end.
# 3. At each step, compare the current element `nums[i]` with the previous
#    element `nums[i-1]`.
# 4. If `nums[i]` is a unique element (i.e., it is different from the
#    element just before it, `nums[i-1]`), then it should be kept.
# 5. Write this unique element to the position of the 'write' pointer:
#    `nums[k] = nums[i]`.
# 6. After writing, increment `k` to prepare for the next unique element.
# 7. If `nums[i]` is a duplicate, we simply do nothing and continue
#    to the next element, effectively skipping it.
# 8. The loop terminates when the 'read' pointer reaches the end of the array.
# 9. The final value of `k` is the total count of unique elements, and the
#    first `k` elements of the `nums` array will contain the unique elements.
#
# Time Complexity: O(n), as both pointers traverse the array at most once.
# Space Complexity: O(1), as the modification is done in-place without
#                    using any additional data structures.

class Solution:
    def removeDuplicates(self, nums: list[int]) -> int:
        n = len(nums)
        if n == 0:
            return 0
        
        # 'k' is the write pointer, starting at index 1
        k = 1
        
        # 'i' is the read pointer, starting at index 1
        for i in range(1, n):
            # Check for a unique element
            if nums[i] != nums[i - 1]:
                nums[k] = nums[i]
                k += 1
                
        return k

# --- Test Cases ---
if __name__ == "__main__":
    sol = Solution()

    # Example 1
    nums1 = [1, 1, 2]
    expected_nums1 = [1, 2]
    k1 = sol.removeDuplicates(nums1)
    print(f"Input: {nums1}, k: {k1}, Modified nums: {nums1[:k1]}, Expected: {expected_nums1}")
    assert k1 == len(expected_nums1)
    assert nums1[:k1] == expected_nums1
    print("-" * 20)

    # Example 2
    nums2 = [0, 0, 1, 1, 1, 2, 2, 3, 3, 4]
    expected_nums2 = [0, 1, 2, 3, 4]
    k2 = sol.removeDuplicates(nums2)
    print(f"Input: {nums2}, k: {k2}, Modified nums: {nums2[:k2]}, Expected: {expected_nums2}")
    assert k2 == len(expected_nums2)
    assert nums2[:k2] == expected_nums2
    print("-" * 20)

    # Edge case: All elements are unique
    nums3 = [1, 2, 3, 4, 5]
    expected_nums3 = [1, 2, 3, 4, 5]
    k3 = sol.removeDuplicates(nums3)
    print(f"Input: {nums3}, k: {k3}, Modified nums: {nums3[:k3]}, Expected: {expected_nums3}")
    assert k3 == len(expected_nums3)
    assert nums3[:k3] == expected_nums3
    print("-" * 20)

    # Edge case: All elements are the same
    nums4 = [7, 7, 7, 7, 7]
    expected_nums4 = [7]
    k4 = sol.removeDuplicates(nums4)
    print(f"Input: {nums4}, k: {k4}, Modified nums: {nums4[:k4]}, Expected: {expected_nums4}")
    assert k4 == len(expected_nums4)
    assert nums4[:k4] == expected_nums4
    print("-" * 20)

    # Edge case: Empty array
    nums5 = []
    expected_nums5 = []
    k5 = sol.removeDuplicates(nums5)
    print(f"Input: {nums5}, k: {k5}, Modified nums: {nums5[:k5]}, Expected: {expected_nums5}")
    assert k5 == len(expected_nums5)
    assert nums5[:k5] == expected_nums5
    print("-" * 20)
````
