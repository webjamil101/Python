# 3330. Find the Original Typed String I

# Easy

Alice is attempting to type a specific string on her computer. However, she tends to be clumsy and may press a key for too long, resulting in a character being typed multiple times.

Although Alice tried to focus on her typing, she is aware that she may still have done this at most once.

You are given a string word, which represents the final output displayed on Alice's screen.

Return the total number of possible original strings that Alice might have intended to type.

# Example 1:

```
Input: word = "abbcccc"

Output: 5

Explanation:

The possible strings are: "abbcccc", "abbccc", "abbcc", "abbc", and "abcccc".
```

# Example 2:

```
Input: word = "abcd"

Output: 1

Explanation:

The only possible string is "abcd".
```

# Example 3:

```
Input: word = "aaaa"

Output: 4
```

# Constraints:

- 1 <= word.length <= 100
- word consists only of lowercase English letters.


````python
class Solution:
    def possibleStringCount(self, word: str) -> int:
        unique_variants = set()
        unique_variants.add(word)
        i = 0

        while i < len(word):
            j = i
            while j < len(word) and word[i] == word[j]:
                j += 1

            duplicate_length = j - i

            if duplicate_length > 1:
                for k in range(1, duplicate_length + 1):
                    variant = word[:i + k] + word[j:]
                    unique_variants.add(variant)

            i = j

        return len(unique_variants)
def test_possible_string_count():
    s = Solution()

    # Basic case with duplicate characters
    assert s.possibleStringCount("aabbcc") == 7
    # Explanation: Original + variants removing 1 or more from each group

    # No duplicates
    assert s.possibleStringCount("abc") == 1

    # All same letters
    assert s.possibleStringCount("aaaa") == 5
    # Variants by removing 1, 2, 3, or all 4 (only adds unique shorter versions)

    # Empty string
    assert s.possibleStringCount("") == 1  # Just the empty string

    # One character
    assert s.possibleStringCount("x") == 1

    # Mixed cases
    assert s.possibleStringCount("aabbaa") == 9

    print("All test cases passed!")

test_possible_string_count()
````

````python
class Solution:
    def possibleStringCount(self, word: str) -> int:
        variants = set()
        variants.add(word)

        i = 0
        while i < len(word):
            j = i
            # Move j to the end of the group of repeated characters
            while j < len(word) and word[j] == word[i]:
                j += 1

            group_length = j - i

            # Try removing some characters from the repeated group
            if group_length > 1:
                for removed in range(1, group_length):
                    new_word = word[:i + removed] + word[j:]
                    variants.add(new_word)

            i = j  # Move to the next group

        return len(variants)

# ðŸ§ª Test cases
def test_possible_string_count():
    s = Solution()

    print("Testing 'aabbcc' âžœ", s.possibleStringCount("aabbcc"))  # Expected: 7
    print("Testing 'abc' âžœ", s.possibleStringCount("abc"))        # Expected: 1
    print("Testing 'aaaa' âžœ", s.possibleStringCount("aaaa"))      # Expected: 5
    print("Testing '' âžœ", s.possibleStringCount(""))              # Expected: 1
    print("Testing 'x' âžœ", s.possibleStringCount("x"))            # Expected: 1
    print("Testing 'aabbaa' âžœ", s.possibleStringCount("aabbaa"))  # Expected: 9

test_possible_string_count()
````
