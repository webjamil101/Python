# 217. Contains Duplicate

**Easy**

> Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

# Example 1:

```Python
Input: nums = [1,2,3,1]

Output: true
```

**Explanation:**

The element 1 occurs at the indices 0 and 3.

# Example 2:

```Python
Input: nums = [1,2,3,4]

Output: false
```

**Explanation:**

All elements are distinct.

# Example 3:

```Python
Input: nums = [1,1,1,3,3,4,3,2,4,2]

Output: true
```

# Constraints:

- 1 <= nums.length <= 105
- -109 <= nums[i] <= 109


````python
class Solution:
    def containsDuplicate(self, nums: list[int]) -> bool:
        seen = set()  # Initialize an empty set
        for num in nums:
            if num in seen:  # Check if the element is already in the set
                return True
            seen.add(num)  # Add the element to the set
        return False
````

````python
class Solution:
    def containsDuplicate(self, nums: list[int]) -> bool:
        nums.sort()  # Sort the array
        for i in range(len(nums) - 1):
            if nums[i] == nums[i+1]:  # Check for adjacent duplicates
                return True
        return False
````

````python
class Solution:
    def containsDuplicate(self, nums: list[int]) -> bool:
        seen_map = {}  # Initialize an empty dictionary
        for num in nums:
            if num in seen_map:  # Check if the element is a key in the dictionary
                return True
            seen_map[num] = True  # Add the element as a key (value doesn't matter)
        return False
````

````python
class Solution:
    def containsDuplicate(self, nums: list[int]) -> bool:
        n = len(nums)
        for i in range(n):
            for j in range(i + 1, n):  # Compare with elements after current
                if nums[i] == nums[j]:
                    return True
        return False
````

````python
from collections import Counter

class Solution:
    def containsDuplicate(self, nums: list[int]) -> bool:
        counts = Counter(nums)  # Count occurrences of all elements
        for count in counts.values():
            if count > 1:
                return True
        return False
````

````python

from collections import Counter

class Solution:
    def containsDuplicate(self, nums: list[int]) -> bool:
        counts = Counter(nums)
        return any(v > 1 for v in counts.values())

# Test Suite
if __name__ == "__main__":
    sol = Solution()

    # âœ… Common Examples
    print(sol.containsDuplicate([1, 2, 3, 1]))                  # True
    print(sol.containsDuplicate([1, 2, 3, 4]))                  # False
    print(sol.containsDuplicate([1,1,1,3,3,4,3,2,4,2]))         # True

    # ðŸš§ Edge Cases
    print(sol.containsDuplicate([]))                           # False â€“ Empty list
    print(sol.containsDuplicate([5]))                          # False â€“ Single element
    print(sol.containsDuplicate([-10**9, 10**9]))              # False â€“ Large extreme values, distinct
    print(sol.containsDuplicate([99] * 100_000))               # True â€“ Large list with duplicates
    print(sol.containsDuplicate(list(range(100_000))))         # False â€“ Large list with all unique elements
````
