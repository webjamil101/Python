2311. Longest Binary Subsequence Less Than or Equal to K

Medium

    You are given a binary string s and a positive integer k.

Return the length of the longest subsequence of s that makes up a binary number less than or equal to k.

Note:

The subsequence can contain leading zeroes.
The empty string is considered to be equal to 0.
A subsequence is a string that can be derived from another string by deleting some or no characters without changing the order of the remaining characters.

Example 1:

Input: s = "1001010", k = 5
Output: 5
Explanation: The longest subsequence of s that makes up a binary number less than or equal to 5 is "00010", as this number is equal to 2 in decimal.
Note that "00100" and "00101" are also possible, which are equal to 4 and 5 in decimal, respectively.
The length of this subsequence is 5, so 5 is returned.
Example 2:

Input: s = "00101001", k = 1
Output: 6
Explanation: "000001" is the longest subsequence of s that makes up a binary number less than or equal to 1, as this number is equal to 1 in decimal.
The length of this subsequence is 6, so 6 is returned.

Constraints:

1 <= s.length <= 1000
s[i] is either '0' or '1'.
1 <= k <= 109


```python
class Solution:
    def longestSubsequence(self, binary_string: str, max_value: int) -> int:
        total_value = 0
        subseq_length = 0
        max_bits = max_value.bit_length()

        for bit_position, char in enumerate(reversed(binary_string)):
            if char == "1":
                if bit_position < max_bits and total_value + (1 << bit_position) <= max_value:
                    total_value += 1 << bit_position
                    subseq_length += 1
            else:
                subseq_length += 1

        return subseq_length

# Test cases
if __name__ == "__main__":
    sol = Solution()

    # ✅ Basic test
    print("Test 1:", sol.longestSubsequence("1001010", 5))  # Expected: 5

    # ✅ All zeros (should return full length)
    print("Test 2:", sol.longestSubsequence("0000000", 1))  # Expected: 7

    # ✅ All ones, small k
    print("Test 3:", sol.longestSubsequence("11111", 1))    # Expected: 1

    # ✅ k is very large
    print("Test 4:", sol.longestSubsequence("1010101010", 10**9))  # Expected: 10

    # ✅ k is 0 (only zeros allowed)
    print("Test 5:", sol.longestSubsequence("101010", 0))   # Expected: 3

    # ✅ Empty string
    print("Test 6:", sol.longestSubsequence("", 10))        # Expected: 0

    # ✅ Single character
    print("Test 7:", sol.longestSubsequence("1", 1))        # Expected: 1
    print("Test 8:", sol.longestSubsequence("1", 0))        # Expected: 0
    print("Test 9:", sol.longestSubsequence("0", 0))        # Expected: 1
```

```python
class Solution:
    def longestSubsequence(self, binary_string: str, max_value: int) -> int:
        total = 0
        length = 0
        max_bits = max_value.bit_length()

        for position, char in enumerate(reversed(binary_string)):
            if char == "1":
                if position < max_bits and total + (1 << position) <= max_value:
                    total += 1 << position
                    length += 1
            else:
                length += 1

        return length
```
