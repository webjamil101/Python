**1935. Maximum Number of Words You Can Type**

**Easy**

**Companies**

There is a malfunctioning keyboard where some letter keys do not work. All other keys on the keyboard work properly.

Given a string text of words separated by a single space (no leading or trailing spaces) and a string brokenLetters of all distinct letter keys that are broken, return the number of words in text you can fully type using this keyboard.

 

**Example 1:**
```python
Input: text = "hello world", brokenLetters = "ad"
Output: 1
```
**Explanation:** We cannot type "world" because the 'd' key is broken.

**Example 2:**
```python
Input: text = "leet code", brokenLetters = "lt"
Output: 1
```
**Explanation:** We cannot type "leet" because the 'l' and 't' keys are broken.

**Example 3:**
```python
Input: text = "leet code", brokenLetters = "e"
Output: 0
```
**Explanation:** We cannot type either word because the 'e' key is broken.
 

**Constraints:**

- 1 <= text.length <= 104
- 0 <= brokenLetters.length <= 26
- text consists of words separated by a single space without any leading or - trailing spaces.
- Each word only consists of lowercase English letters.
- brokenLetters consists of distinct lowercase English letters.

````python
class Solution:
    def canBeTypedWords(self, text: str, brokenLetters: str) -> int:
        # Step 1: Create a set for fast lookup of broken letters
        broken = set(brokenLetters)
        
        # Step 2: Split the text into individual words
        words = text.split()
        
        # Step 3: Initialize counter for typable words
        count = 0
        
        # Step 4: Iterate over words
        for word in words:
            # Check if word contains any broken letter
            if all(char not in broken for char in word):
                count += 1  # Word can be typed
        
        return count
````

````python
class Solution:
    def canBeTypedWords(self, text: str, brokenLetters: str) -> int:
        words = text.split()
        count = 0
        
        for word in words:
            is_broken = False
            for char in word:
                if char in brokenLetters:  # Linear search in brokenLetters
                    is_broken = True
                    break
            if not is_broken:
                count += 1
        
        return count
````

````python
import re

class Solution:
    def canBeTypedWords(self, text: str, brokenLetters: str) -> int:
        # Compile regex to match any broken character
        pattern = re.compile(f"[{re.escape(brokenLetters)}]")
        
        count = 0
        for word in text.split():
            # If pattern doesn't match any char in word, it can be typed
            if not pattern.search(word):
                count += 1
        
        return count
````

````python
class Solution:
    def canBeTypedWords(self, text: str, brokenLetters: str) -> int:
        # Create bitmask for broken letters
        broken_mask = 0
        for ch in brokenLetters:
            broken_mask |= 1 << (ord(ch) - ord('a'))
        
        count = 0
        for word in text.split():
            word_mask = 0
            for ch in word:
                word_mask |= 1 << (ord(ch) - ord('a'))
            if word_mask & broken_mask == 0:
                count += 1
        
        return count
````
