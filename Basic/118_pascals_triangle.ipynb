# 118. Pascal's Triangle

**Easy**

Given an integer numRows, return the first numRows of Pascal's triangle.

In Pascal's triangle, each number is the sum of the two numbers directly above it as shown:

# Example 1:

```python
Input: numRows = 5
Output: [[1],[1,1],[1,2,1],[1,3,3,1],[1,4,6,4,1]]
```

# Example 2:

```python
Input: numRows = 1
Output: [[1]]

```

**Constraints**:

- 1 <= numRows <= 30


````python
class Solution:
    def generate(self, numRows: int) -> list[list[int]]:
        if numRows == 0:
            return []

        triangle = [[1]]
        for i in range(1, numRows):
            prev_row = triangle[-1]
            new_row = [1]
            for j in range(1, len(prev_row)):
                new_row.append(prev_row[j-1] + prev_row[j])
            new_row.append(1)
            triangle.append(new_row)
        
        return triangle
````

````python
class Solution:
    def generate(self, numRows: int) -> list[list[int]]:
        """
        Generates the first numRows of Pascal's triangle using an iterative approach.
        """
        result = []
        for i in range(numRows):
            # Each row is a list. Initialize the current row with 1s.
            # The length of the i-th row is i+1.
            current_row = [1] * (i + 1)

            # The inner elements are the sum of the two elements from the previous row.
            # We start from the second element (index 1) and go up to the second-to-last element.
            for j in range(1, i):
                current_row[j] = result[i - 1][j - 1] + result[i - 1][j]
            
            result.append(current_row)
        
        return result
````

````python
class Solution:
    def getRow(self, line: int) -> list[int]:
        """
        Helper function to generate a single row of Pascal's triangle
        using the combination formula C(n, k) = C(n, k-1) * (n-k+1) / k.
        """
        prev = 1
        temp = [1]
        
        # The loop calculates each element based on the previous one in the same row.
        for i in range(1, line + 1):
            curr = (line - i + 1) * prev // i
            temp.append(curr)
            prev = curr
        
        return temp

    def generate(self, numRows: int) -> list[list[int]]:
        """
        Generates the first numRows of Pascal's triangle by calling a
        helper function for each row.
        """
        if numRows == 0:
            return []

        result = []
        for lineNo in range(numRows):
            result.append(self.getRow(lineNo))
        
        return result
````

````python
class Solution:
    def generate(self, numRows: int) -> list[list[int]]:
        """
        Generates the first numRows of Pascal's triangle using an iterative approach.
        """
        if numRows <= 0:
            return []  # Edge case: no rows requested

        result = []
        for i in range(numRows):
            current_row = [1] * (i + 1)
            for j in range(1, i):
                current_row[j] = result[i - 1][j - 1] + result[i - 1][j]
            result.append(current_row)
        return result


def print_triangle(triangle: list[list[int]]):
    for row in triangle:
        print(row)


def main():
    sol = Solution()

    # ðŸ§ª Test Case 1: Basic case
    print("Test Case 1: numRows = 5")
    print_triangle(sol.generate(5))
    print()

    # ðŸ§ª Test Case 2: Edge case - numRows = 0
    print("Test Case 2: numRows = 0")
    print_triangle(sol.generate(0))
    print()

    # ðŸ§ª Test Case 3: Single row
    print("Test Case 3: numRows = 1")
    print_triangle(sol.generate(1))
    print()

    # ðŸ§ª Test Case 4: Larger triangle
    print("Test Case 4: numRows = 10")
    print_triangle(sol.generate(10))
    print()

    # ðŸ§ª Test Case 5: Negative input
    print("Test Case 5: numRows = -3")
    print_triangle(sol.generate(-3))
    print()


if __name__ == "__main__":
    main()
````
