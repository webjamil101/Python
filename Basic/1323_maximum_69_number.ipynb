**1323. Maximum 69 Number**

**Easy**

**Companies:oracle**

You are given a positive integer num consisting only of digits 6 and 9.

Return the maximum number you can get by changing at most one digit (6 becomes 9, and 9 becomes 6).

**Example 1:**

```python
Input: num = 9669
Output: 9969
```

**Explanation:**

- Changing the first digit results in 6669.
- Changing the second digit results in 9969.
- Changing the third digit results in 9699.
- Changing the fourth digit results in 9666.
- The maximum number is 9969.

**Example 2:**

```python
Input: num = 9996
Output: 9999
```

**Explanation:** Changing the last digit 6 to 9 results in the maximum number.

**Example 3:**

```python
Input: num = 9999
Output: 9999
```

**Explanation:** It is better not to apply any change.

**Constraints:**

1 <= num <= 104
num consists of only 6 and 9 digits.


````python
class Solution:
    def maximum69Number (self, num: int) -> int:
        # Algorithm:
        # 1. Convert the integer to a string to easily access and modify its digits.
        # 2. Iterate through the string from left to right.
        # 3. Find the first occurrence of '6'.
        # 4. Change it to '9' and return the new number.
        # 5. If no '6' is found, the number is already maximized, so return the original number.

        s = str(num)
        # s.find('6') returns the index of the first '6', or -1 if not found.
        first_six_index = s.find('6')

        if first_six_index != -1:
            # Construct the new string by replacing the first '6'.
            s_list = list(s)
            s_list[first_six_index] = '9'
            return int("".join(s_list))
        
        # If no '6' exists, the number is already the maximum.
        return num
````

````python
class Solution:
    def maximum69Number(self, num: int) -> int:
        # Algorithm:
        # 1. Iterate through the digits from right to left using modulo and division.
        # 2. Keep track of the position of the first '6' encountered (which is the
        #    leftmost '6' in the number).
        # 3. Use the place value of this '6' to add 3 to the number, effectively
        #    changing the '6' to a '9'.
        
        temp_num = num
        power_of_10 = 1
        first_six_pos = 0

        # Loop from right to left to find the leftmost '6'
        while temp_num > 0:
            digit = temp_num % 10
            if digit == 6:
                first_six_pos = power_of_10  # Save the place value of the leftmost '6'
            
            temp_num //= 10
            power_of_10 *= 10
        
        # If a '6' was found, add 3 times its place value to the original number.
        if first_six_pos != 0:
            return num + 3 * first_six_pos
        else:
            return num
sol = Solution()

# ✅ Basic case
print(sol.maximum69Number(9669))  # Expected: 9969

# ✅ Only one digit
print(sol.maximum69Number(6))     # Expected: 9
print(sol.maximum69Number(9))     # Expected: 9

# ✅ Multiple 6s, only the leftmost should change
print(sol.maximum69Number(6699))  # Expected: 9699

# ✅ No 6s
print(sol.maximum69Number(9999))  # Expected: 9999

# ✅ 6 at the end
print(sol.maximum69Number(9996))  # Expected: 9999

# ✅ 6 at the beginning
print(sol.maximum69Number(6999))  # Expected: 9999

# ✅ Large number
print(sol.maximum69Number(696969))  # Expected: 996969
````

````python
class Solution:
    def maximum69Number(self, num: int) -> int:
        place = 0
        index = -1
        temp = num

        while temp > 0:
            remain = temp % 10
            if remain == 6:
                index = place
            temp //= 10
            place += 1

        if index == -1:
            return num

        return num + 3 * (10 ** index)
sol = Solution()
print(sol.maximum69Number(9669))  # Output: 9969
print(sol.maximum69Number(9999))  # Output: 9999
print(sol.maximum69Number(6666))  # Output: 6696
print(sol.maximum69Number(6969))  # Output: 9969
print(sol.maximum69Number(6))     # Output: 9
print(sol.maximum69Number(9))     # Output: 9
print(sol.maximum69Number(6699))  # Output: 9699
````
