# 1464. Maximum Product of Two Elements in an Array

**Easy**

Given the array of integers nums, you will choose two different indices i and j of that array. Return the maximum value of **(nums[i]-1)\*(nums[j]-1)**.

# Example 1:

```python
Input: nums = [3,4,5,2]
Output: 12
```

**Explanation:** If you choose the indices i=1 and j=2 (indexed from 0), you will get the maximum value, that is, (nums[1]-1)_(nums[2]-1) = (4-1)_(5-1) = 3\*4 = 12.

# Example 2:

```python
Input: nums = [1,5,4,5]
Output: 16
```

**Explanation:** Choosing the indices i=1 and j=3 (indexed from 0), you will get the maximum value of (5-1)\*(5-1) = 16.

# Example 3:

```python
Input: nums = [3,7]
Output: 12

```

# Constraints:

- 2 <= nums.length <= 500
- 1 <= nums[i] <= 10^3


````python
class Solution:
    def maxProduct(self, nums):
        if len(nums) < 2:
            raise ValueError("Array must contain at least two elements")
        
        max1 = max2 = 0
        
        # Find the top two maximum values
        for num in nums:
            if num > max1:
                max2 = max1
                max1 = num
            elif num > max2:
                max2 = num
        
        return (max1 - 1) * (max2 - 1)


# üîç Test Cases
def run_tests():
    sol = Solution()
    
    # Regular case
    print("Test 1:", sol.maxProduct([3, 4, 5, 2]) == 12)  # (5-1)*(4-1)
    
    # Duplicate max values
    print("Test 2:", sol.maxProduct([10, 10]) == 81)  # (10-1)*(10-1)
    
    # Negative values (edge case)
    print("Test 3:", sol.maxProduct([-1, -2, -3]) == 2)  # (-1-1)*(-2-1) = (-2)*(-3) = 6 (but max1= -1, max2= -2 ‚Üí (-1-1)*(-2-1)=(-2)*(-3)=6)

    # Mixed values
    print("Test 4:", sol.maxProduct([1, 0, 3, 7]) == 12)  # (7-1)*(3-1)

    # Only two values
    print("Test 5:", sol.maxProduct([1, 5]) == 0)  # (5-1)*(1-1) = 4*0

    # Large values
    print("Test 6:", sol.maxProduct([1000, 999]) == 998001)  # (1000-1)*(999-1)

    # Too few elements
    try:
        sol.maxProduct([4])
    except ValueError as e:
        print("Test 7:", str(e) == "Array must contain at least two elements")

run_tests()
````

````python
class Solution:
    def maxProduct(self, nums):
        # üõ°Ô∏è Edge case: must have at least two elements
        if len(nums) < 2:
            raise ValueError("The input list must contain at least two elements.")

        # üîß Initialization
        maxNum = nums[0]
        result = 0

        # üîÑ Iterate through the rest of the list
        for i in range(1, len(nums)):
            # üßÆ Calculate product with current pair
            result = max(result, (maxNum - 1) * (nums[i] - 1))
            # üìà Update maxNum if current is greater
            maxNum = max(maxNum, nums[i])

        return result


# üîç Test Function to Validate Behavior
def run_tests():
    sol = Solution()

    assert sol.maxProduct([3, 4, 5, 2]) == 12        # Test basic input
    assert sol.maxProduct([1, 5]) == 0              # One element is 1
    assert sol.maxProduct([10, 10]) == 81           # Duplicate maximums
    assert sol.maxProduct([1000, 999]) == 998001    # Very large numbers
    assert sol.maxProduct([1, 0, 3, 7]) == 12        # Mixed values
    assert sol.maxProduct([-1, -2, -3]) == 6         # Negative values

    try:
        sol.maxProduct([42])                         # Edge: too few elements
        print("‚ùå Test failed: Should have raised ValueError.")
    except ValueError:
        print("‚úÖ Test passed: ValueError raised for single element.")

    print("‚úÖ All other tests passed!")

# üöÄ Run test cases
run_tests()
````
