# 1716. Calculate Money in Leetcode Bank

# Easy

Hercy wants to save money for his first car. He puts money in the Leetcode bank every day.

He starts by putting in $1 on Monday, the first day. Every day from Tuesday to Sunday, he will put in $1 more than the day before. On every subsequent Monday, he will put in $1 more than the previous Monday.

Given n, return the total amount of money he will have in the Leetcode bank at the end of the nth day.

# Example 1:

```
Input: n = 4
Output: 10
Explanation: After the 4th day, the total is 1 + 2 + 3 + 4 = 10.
```

# Example 2:

```
Input: n = 10
Output: 37
Explanation: After the 10th day, the total is (1 + 2 + 3 + 4 + 5 + 6 + 7) + (2 + 3 + 4) = 37. Notice that on the 2nd Monday, Hercy only puts in $2.
```

# Example 3:

```
Input: n = 20
Output: 96
Explanation: After the 20th day, the total is (1 + 2 + 3 + 4 + 5 + 6 + 7) + (2 + 3 + 4 + 5 + 6 + 7 + 8) + (3 + 4 + 5 + 6 + 7 + 8) = 96.

```

# Constraints:

- 1 <= n <= 1000


````python
class Solution:
    def calculateMoney_simulation(self, n: int) -> int:
        total_money = 0
        current_deposit = 1
        day_of_week = 1  # 1 for Monday, 7 for Sunday
        monday_start_deposit = 1

        for day in range(1, n + 1):
            total_money += current_deposit
            current_deposit += 1
            day_of_week += 1

            if day_of_week > 7:  # End of the week (finished Sunday)
                day_of_week = 1  # Reset to Monday
                monday_start_deposit += 1  # Next Monday starts with $1 more
                current_deposit = monday_start_deposit  # Set deposit for new Monday

        return total_money

# Test cases
def run_tests():
    solution = Solution()
    
    # Edge cases
    assert solution.calculateMoney_simulation(1) == 1  # First day deposit
    assert solution.calculateMoney_simulation(7) == 28  # Full week
    assert solution.calculateMoney_simulation(14) == 77  # Two full weeks

    # Typical cases
    assert solution.calculateMoney_simulation(10) == 37  # Partial second week
    assert solution.calculateMoney_simulation(20) == 96  # Three weeks
    
    # Large input
    assert solution.calculateMoney_simulation(100) == 2300  # Large value test
    
    print("All test cases passed!")

# Run the tests
run_tests()
````
