**21. Merge Two Sorted Lists**

**Easy**

**Companies**: Amazon Google Microsoft


You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists into one sorted list. The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.

 

**Example 1:**

```python
Input: list1 = [1,2,4], list2 = [1,3,4]
Output: [1,1,2,3,4,4]
```
**Example 2:**
```python
Input: list1 = [], list2 = []
Output: []
```
**Example 3:**
```python
Input: list1 = [], list2 = [0]
Output: [0]
 ```

**Constraints:**

- The number of nodes in both lists is in the range [0, 50].
- -100 <= Node.val <= 100
- Both list1 and list2 are sorted in non-decreasing order.

````python
# --------------------------------------------------------
# Approach 1: Iterative using dummy node
# --------------------------------------------------------
# Algorithm:
# 1. Create dummy node to simplify edge cases.
# 2. Use 'current' pointer to build the merged list.
# 3. Compare nodes of list1 and list2 iteratively and attach smaller to current.next.
# 4. Attach the remaining nodes after one list is exhausted.
# 5. Return dummy.next as the head of merged list.
# --------------------------------------------------------

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def mergeTwoLists(self, list1: ListNode, list2: ListNode) -> ListNode:
        dummy = ListNode()
        current = dummy

        while list1 and list2:
            if list1.val < list2.val:
                current.next = list1
                list1 = list1.next
            else:
                current.next = list2
                list2 = list2.next
            current = current.next

        # Attach the remaining nodes
        current.next = list1 if list1 else list2

        return dummy.next
````

````python
# --------------------------------------------------------
# Approach 2: Recursive
# --------------------------------------------------------
# Algorithm:
# 1. Base case: if one list is None, return the other.
# 2. Compare the values of current nodes.
# 3. Attach the smaller node to the result of recursive call.
# 4. Return the smaller node as new head.
# --------------------------------------------------------

class Solution:
    def mergeTwoLists(self, list1: ListNode, list2: ListNode) -> ListNode:
        if not list1:
            return list2
        if not list2:
            return list1

        if list1.val < list2.val:
            list1.next = self.mergeTwoLists(list1.next, list2)
            return list1
        else:
            list2.next = self.mergeTwoLists(list1, list2.next)
            return list2
````

````python
# --------------------------------------------------------
# Approach 3: Create a new list
# --------------------------------------------------------
# Algorithm:
# 1. Create dummy node for new list.
# 2. Compare list1 and list2 values, append new nodes to current.
# 3. Attach remaining nodes as new nodes.
# --------------------------------------------------------

class Solution:
    def mergeTwoLists(self, list1: ListNode, list2: ListNode) -> ListNode:
        dummy = ListNode()
        current = dummy

        while list1 and list2:
            if list1.val < list2.val:
                current.next = ListNode(list1.val)
                list1 = list1.next
            else:
                current.next = ListNode(list2.val)
                list2 = list2.next
            current = current.next

        # Append remaining nodes as new nodes
        while list1:
            current.next = ListNode(list1.val)
            list1 = list1.next
            current = current.next
        while list2:
            current.next = ListNode(list2.val)
            list2 = list2.next
            current = current.next

        return dummy.next
````
