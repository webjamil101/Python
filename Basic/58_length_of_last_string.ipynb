**58. Length of Last Word**

**Easy**

**Companies**:

Given a string s consisting of words and spaces, return the length of the last word in the string.

A word is a maximal substring consisting of non-space characters only.

**Example 1**

```python
Input: s = "Hello World"
Output: 5
```

**Explanation:** The last word is "World" with length 5.

**Example 2:**

```python
Input: s = " fly me to the moon "
Output: 4
```

**Explanation:** The last word is "moon" with length 4.
**Example 3:**

```python
Input: s = "luffy is still joyboy"
Output: 6
```

**Explanation:** The last word is "joyboy" with length 6.

**Constraints:**

- 1 <= s.length <= 104
- s consists of only English letters and spaces ' '.
- There will be at least one word in s.


````python
# =====================================================
# ðŸ§  Approach 1: Using split()
# -----------------------------------------------------
# ðŸ”¹ Algorithm:
#   1. Use Python's built-in split() â†’ splits on whitespace.
#   2. It automatically ignores extra spaces.
#   3. Return the length of the last element of the list.
#
# ðŸ”¹ Example:
#   s = "   fly me   to   the moon  "
#   s.split() = ['fly', 'me', 'to', 'the', 'moon']
#   â†’ last word = 'moon' â†’ len = 4
#
# ðŸ”¹ Time Complexity: O(n)
# ðŸ”¹ Space Complexity: O(n)
# =====================================================
class Solution:
    def lengthOfLastWord(self, s: str) -> int:
        words = s.split()
        return len(words[-1])
````

````python
# =====================================================
# ðŸ§  Approach 2: Using rsplit(maxsplit=1)
# -----------------------------------------------------
# ðŸ”¹ Algorithm:
#   1. Remove extra spaces using strip().
#   2. Use rsplit(' ', 1) to split only once from the right.
#   3. Return the length of the last segment.
#
# ðŸ”¹ Example:
#   s = "Hello World"
#   s.strip().rsplit(' ', 1) = ['Hello', 'World']
#   â†’ last = 'World' â†’ len = 5
#
# ðŸ”¹ Time Complexity: O(n)
# ðŸ”¹ Space Complexity: O(k) (k = length of last word)
# =====================================================
class Solution:
    def lengthOfLastWord(self, s: str) -> int:
        return len(s.strip().rsplit(' ', 1)[-1])
````

````python
# =====================================================
# ðŸ§  Approach 3: Manual Reverse Traversal
# -----------------------------------------------------
# ðŸ”¹ Algorithm:
#   1. Start from the end of string.
#   2. Skip all trailing spaces.
#   3. Count characters until another space or start of string.
#
# ðŸ”¹ Example:
#   s = "   fly me   to   the moon  "
#   Traverse from right:
#     - skip trailing spaces
#     - count 'moon' â†’ length = 4
#
# ðŸ”¹ Time Complexity: O(n)
# ðŸ”¹ Space Complexity: O(1)
# ðŸ”¹ Best for interviews â€” optimal and simple.
# =====================================================
class Solution:
    def lengthOfLastWord(self, s: str) -> int:
        i = len(s) - 1
        length = 0

        # Step 1: skip trailing spaces
        while i >= 0 and s[i] == ' ':
            i -= 1

        # Step 2: count characters of last word
        while i >= 0 and s[i] != ' ':
            length += 1
            i -= 1

        return length
````

````python
# =====================================================
# ðŸ§  Approach 4: Using rfind()
# -----------------------------------------------------
# ðŸ”¹ Algorithm:
#   1. Remove trailing spaces using rstrip().
#   2. Find last space index using rfind(' ').
#   3. Subtract that index from total length to get last word length.
#
# ðŸ”¹ Example:
#   s = "Hello World"
#   s.rstrip() = "Hello World"
#   last_space = 5 â†’ len = 11 - 5 - 1 = 5
#
# ðŸ”¹ Time Complexity: O(n)
# ðŸ”¹ Space Complexity: O(1)
# =====================================================
class Solution:
    def lengthOfLastWord(self, s: str) -> int:
        s = s.rstrip()
        last_space = s.rfind(' ')
        return len(s) - last_space - 1
````

````python
# =====================================================
# ðŸ§  Approach 5: Using Regular Expressions
# -----------------------------------------------------
# ðŸ”¹ Algorithm:
#   1. Use regex `re.findall(r'\b\w+\b', s)` to extract all words.
#   2. Return length of the last one.
#
# ðŸ”¹ Example:
#   s = "fly me to the moon"
#   re.findall(...) = ['fly', 'me', 'to', 'the', 'moon']
#   â†’ last = 'moon' â†’ len = 4
#
# ðŸ”¹ Time Complexity: O(n)
# ðŸ”¹ Space Complexity: O(n)
# =====================================================
import re
class Solution:
    def lengthOfLastWord(self, s: str) -> int:
        words = re.findall(r'\b\w+\b', s)
        return len(words[-1])
````
