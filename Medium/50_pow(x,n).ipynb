**50. Pow(x, n)**

**Medium**

**Companies**: Adobe Amazon Apple Asana Bloomberg Facebook Goldman Sachs Google LinkedIn Microsoft Morgan Stanley Oracle Uber Walmart Labs

Implement pow(x, n), which calculates x raised to the power n (i.e., xn).

**Example 1:**

```
Input: x = 2.00000, n = 10
Output: 1024.00000
```

**Example 2:**

```
Input: x = 2.10000, n = 3
Output: 9.26100
```

**Example 3:**

```
Input: x = 2.00000, n = -2
Output: 0.25000
```

**Explanation:** 2-2 = 1/22 = 1/4 = 0.25

**Constraints:**

- -100.0 < x < 100.0
- -231 <= n <= 231-1
- n is an integer.
- Either x is not zero or n > 0.
- -104 <= xn <= 104


````python
class Solution:
    def myPow(self, x: float, n: int) -> float:
        """
        Algorithm (Recursive Binary Exponentiation):
        1. Base cases:
           - n == 0 â†’ return 1
        2. Recursively compute:
           - x^(n//2)
        3. If n is even:
           - result = half * half
        4. If n is odd:
           - result = half * half * x
        5. If n is negative:
           - return 1 / result

        Time Complexity: O(log |n|)
        Space Complexity: O(log |n|) (recursion stack)
        """

        def fastPow(x, n):
            if n == 0:
                return 1.0

            half = fastPow(x, n // 2)

            if n % 2 == 0:
                return half * half
            else:
                return half * half * x

        if n < 0:
            return 1 / fastPow(x, -n)
        return fastPow(x, n)
````

````python
class Solution:
    def myPow(self, x: float, n: int) -> float:
        """
        Algorithm (Iterative Binary Exponentiation):

        1. If n is negative:
           - Replace x with 1/x
           - Convert n to positive.
        2. Initialize result = 1.
        3. While n > 0:
           a) If n is odd:
              - Multiply result by x.
           b) Square x.
           c) Divide n by 2 (integer division).
        4. Return result.

        Time Complexity: O(log |n|)
        Space Complexity: O(1)
        """
        if n == 0:
            return 1.0

        if n < 0:
            x = 1 / x
            n = -n

        result = 1.0

        while n > 0:
            if n % 2 == 1:
                result *= x
            x *= x
            n //= 2

        return result
````
