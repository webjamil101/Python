**2787. Ways to Express an Integer as Sum of Powers**

**Medium**

Companies

Given two positive integers n and x.

Return the number of ways n can be expressed as the sum of the xth power of unique positive integers, in other words, the number of sets of unique integers [n1, n2, ..., nk] where n = n1x + n2x + ... + nkx.

Since the result can be very large, return it modulo 109 + 7.

For example, if n = 160 and x = 3, one way to express n is n = 23 + 33 + 53.

**Example 1:**

```python
Input: n = 10, x = 2
Output: 1
```

**Explanation:** We can express n as the following: n = 32 + 12 = 10.
It can be shown that it is the only way to express 10 as the sum of the 2nd power of unique integers.

**Example 2:**

```python
Input: n = 4, x = 1
Output: 2
```

**Explanation:** We can express n in the following ways:

- n = 41 = 4.
- n = 31 + 11 = 4.

**Constraints**:

- 1 <= n <= 300
- 1 <= x <= 5


````python
import math
from functools import lru_cache

class Solution:
    def numberOfWays(self, n: int, x: int) -> int:
        MOD = 10**9 + 7

        # The @lru_cache decorator handles memoization automatically.
        @lru_cache(None)
        def solve(current_n, num):
            if current_n == 0:
                return 1
            
            if current_n < 0:
                return 0
            
            curr_power_value = int(math.pow(num, x))

            if curr_power_value > current_n:
                return 0

            # Option 1: Take the current power value.
            take = solve(current_n - curr_power_value, num + 1)
            
            # Option 2: Skip the current power value.
            skip = solve(current_n, num + 1)
            
            return (take + skip) % MOD

        return solve(n, 1)
````
