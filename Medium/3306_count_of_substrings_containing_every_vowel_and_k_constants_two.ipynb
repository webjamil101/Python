> 3306. Count of Substrings Containing Every Vowel and K Consonants II

Medium

```
You are given a string word and a non-negative integer k.

Return the total number of substrings of word that contain every vowel ('a', 'e', 'i', 'o', and 'u') at least once and exactly k consonants.

```

```
Example 1:

Input: word = "aeioqq", k = 1

Output: 0

Explanation:

There is no substring with every vowel.

Example 2:

Input: word = "aeiou", k = 0

Output: 1

Explanation:

The only substring with every vowel and zero consonants is word[0..4], which is "aeiou".

Example 3:

Input: word = "ieaouqqieaouqq", k = 1

Output: 3

Explanation:

The substrings with every vowel and one consonant are:

word[0..5], which is "ieaouq".
word[6..11], which is "qieaou".
word[7..12], which is "ieaouq".
```

```
Constraints:



5 <= word.length <= 2 * 105
word consists only of lowercase English letters.
0 <= k <= word.length - 5


```


````python
def count_vowel_substrings(word: str, k: int) -> int:
    from collections import defaultdict

    vowels = set('aeiou')
    n = len(word)
    total = 0

    left = 0
    vowel_count = defaultdict(int)
    unique_vowels = 0
    consonant_count = 0

    for right in range(n):
        char = word[right]
        if char in vowels:
            if vowel_count[char] == 0:
                unique_vowels += 1
            vowel_count[char] += 1
        else:
            consonant_count += 1

        while left <= right and consonant_count > k:
            if word[left] in vowels:
                vowel_count[word[left]] -= 1
                if vowel_count[word[left]] == 0:
                    unique_vowels -= 1
            else:
                consonant_count -= 1
            left += 1

        # Only when we have all vowels and exactly k consonants
        if unique_vowels == 5 and consonant_count == k:
            # Count how many substrings ending at right can be valid
            temp_left = left
            temp_vowel_count = vowel_count.copy()
            temp_consonant_count = consonant_count
            while temp_left <= right and temp_consonant_count == k and len([v for v in vowels if temp_vowel_count[v] > 0]) == 5:
                total += 1
                if word[temp_left] in vowels:
                    temp_vowel_count[word[temp_left]] -= 1
                else:
                    temp_consonant_count -= 1
                temp_left += 1

    return total
````
