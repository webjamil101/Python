**1657. Determine if Two Strings Are Close**

**Medium**

**Companies**: Google, Microsoft, Atlassian, Morgan Stanely, Adobe, Uber

Two strings are considered close if you can attain one from the other using the following operations:

- Operation 1: Swap any two existing characters.
  - For example, abcde -> aecdb
- Operation 2: Transform every occurrence of one existing character into another existing character, and do the same with the other character.
  - For example, aacabb -> bbcbaa (all a's turn into b's, and all b's turn into a's)

You can use the operations on either string as many times as necessary.

Given two strings, word1 and word2, return true if word1 and word2 are close, and false otherwise.

**Example 1:**

```python
Input: word1 = "abc", word2 = "bca"
Output: true
```

**Explanation:** You can attain word2 from word1 in 2 operations.

Apply Operation 1: "abc" -> "acb"

Apply Operation 1: "acb" -> "bca"

**Example 2:**

```python
Input: word1 = "a", word2 = "aa"
Output: false
```

Explanation: It is impossible to attain word2 from word1, or vice versa, in any number of operations.

**Example 3:**

```python
Input: word1 = "cabbba", word2 = "abbccc"
Output: true
```

**Explanation:** You can attain word2 from word1 in 3 operations.

Apply Operation 1: "cabbba" -> "caabbb"

Apply Operation 2: "caabbb" -> "baaccc"

Apply Operation 2: "baaccc" -> "abbccc"

**Constraints:**

- 1 <= word1.length, word2.length <= 105
- word1 and word2 contain only lowercase English letters.


````python
"""
Approach 1: Set Comparison + Sorted Frequencies
------------------------------------------------
Algorithm:
1. If lengths differ → impossible → return False.
2. Count frequency of characters in both strings.
3. Condition A:
      They must contain the SAME SET of characters.
4. Condition B:
      Their frequency lists must match after sorting.
5. If both conditions satisfy → return True.

Reason:
- Operation 1 allows rearranging positions → ignore order.
- Operation 2 allows swapping all frequencies → frequency multiset must match.

Time Complexity:  O(n log n)
Space Complexity: O(1) (only 26 lowercase chars)
"""
from collections import Counter

class Solution:
    def closeStrings(self, word1: str, word2: str) -> bool:
        c1 = Counter(word1)
        c2 = Counter(word2)

        # Condition 1: same characters
        if set(c1.keys()) != set(c2.keys()):
            return False

        # Condition 2: frequency multisets match
        return sorted(c1.values()) == sorted(c2.values())
````

````python
"""
Approach 2: Array Frequency + Sorted Check
------------------------------------------
Algorithm:
1. Create arrays freq1[26], freq2[26].
2. Fill them by counting characters.
3. Check if both strings use the same characters.
4. Sort both frequency arrays.
5. Compare sorted results.

Time Complexity:  O(n + 26 log 26) = O(n)
Space Complexity: O(1)
"""
class Solution2:
    def closeStrings(self, word1: str, word2: str) -> bool:
        if len(word1) != len(word2):
            return False

        freq1 = [0] * 26
        freq2 = [0] * 26

        for c in word1:
            freq1[ord(c) - 97] += 1
        for c in word2:
            freq2[ord(c) - 97] += 1

        # Condition 1: same character presence
        for i in range(26):
            if (freq1[i] == 0) != (freq2[i] == 0):
                return False

        # Condition 2: compare sorted frequencies
        return sorted(freq1) == sorted(freq2)
````

````python
"""
Approach 3: Frequency-of-Frequencies
------------------------------------
Algorithm:
1. Count character frequencies using Counter.
2. Condition A:
      The set of characters must match.
3. Build frequency-of-frequency maps:
      Example:
         freq {a:2,b:3,c:1} -> {1:1, 2:1, 3:1}
4. Compare these maps.
5. If both match → strings are close.

Time Complexity:  O(n + k) = O(n)
Space Complexity: O(k) = O(1)
"""
from collections import Counter

class Solution3:
    def closeStrings(self, word1: str, word2: str) -> bool:
        c1 = Counter(word1)
        c2 = Counter(word2)

        # Condition A: same set of characters
        if set(c1.keys()) != set(c2.keys()):
            return False

        # Condition B: frequency-of-frequencies match
        return Counter(c1.values()) == Counter(c2.values())
````

````python
"""
Approach 4: Optimized Early Exit
--------------------------------
Algorithm:
1. Count freq using Counter.
2. If character sets mismatch → return False.
3. Compare sorted frequency lists.
4. Return result.

(Just an optimized version of Approach 1)

Time Complexity:  O(n log n)
Space Complexity: O(1)
"""
from collections import Counter

class Solution4:
    def closeStrings(self, word1: str, word2: str) -> bool:
        c1, c2 = Counter(word1), Counter(word2)

        # Early stop: characters mismatch
        if c1.keys() != c2.keys():
            return False

        return sorted(c1.values()) == sorted(c2.values())
````
