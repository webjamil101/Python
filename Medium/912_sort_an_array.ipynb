912. Sort an Array

Medium

Given an array of integers nums, sort the array in ascending order and return it.

You must solve the problem without using any built-in functions in O(nlog(n)) time complexity and with the smallest space complexity possible.

# Example 1:

```
Input: nums = [5,2,3,1]
Output: [1,2,3,5]
Explanation: After sorting the array, the positions of some numbers are not changed (for example, 2 and 3), while the positions of other numbers are changed (for example, 1 and 5).
```

# Example 2:

```
Input: nums = [5,1,1,2,0,0]
Output: [0,0,1,1,2,5]
Explanation: Note that the values of nums are not necessairly unique.
```

# Constraints:

- 1 <= nums.length <= 5 \* 104
- -5 _ 104 <= nums[i] <= 5 _ 104


````python
class Solution:
    def sortArray(self, nums: list[int]) -> list[int]:
        def swap(items, l, r):
            temp = items[l]
            items[l] = items[r]
            items[r] = temp

        def partition(items, start, end):
            pivot = items[end]
            s = start
            for i in range(start, end):
                if items[i] <= pivot:
                    swap(items, s, i)
                    s += 1
            swap(items, s, end)
            return s

        def quickSort(items, left, right):
            if left < right:
                pIdx = partition(items, left, right)
                quickSort(items, left, pIdx - 1)
                quickSort(items, pIdx + 1, right)
            return items

        return quickSort(nums, 0, len(nums) - 1)

# --- Edge Cases and Test Cases ---
if __name__ == "__main__":
    solver = Solution()

    # Test Case 1: Empty array
    nums1 = []
    expected1 = []
    result1 = solver.sortArray(nums1)
    print(f"Input: {nums1}, Expected: {expected1}, Result: {result1}, {'Pass' if result1 == expected1 else 'Fail'}")

    # Test Case 2: Single element array
    nums2 = [5]
    expected2 = [5]
    result2 = solver.sortArray(nums2)
    print(f"Input: {nums2}, Expected: {expected2}, Result: {result2}, {'Pass' if result2 == expected2 else 'Fail'}")

    # Test Case 3: Already sorted array
    nums3 = [1, 2, 3, 4, 5]
    expected3 = [1, 2, 3, 4, 5]
    result3 = solver.sortArray(nums3)
    print(f"Input: {nums3}, Expected: {expected3}, Result: {result3}, {'Pass' if result3 == expected3 else 'Fail'}")

    # Test Case 4: Reverse sorted array
    nums4 = [5, 4, 3, 2, 1]
    expected4 = [1, 2, 3, 4, 5]
    result4 = solver.sortArray(nums4)
    print(f"Input: {nums4}, Expected: {expected4}, Result: {result4}, {'Pass' if result4 == expected4 else 'Fail'}")

    # Test Case 5: Array with duplicate elements
    nums5 = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
    expected5 = [1, 1, 2, 3, 3, 4, 5, 5, 5, 6, 9]
    result5 = solver.sortArray(nums5)
    print(f"Input: {nums5}, Expected: {expected5}, Result: {result5}, {'Pass' if result5 == expected5 else 'Fail'}")

    # Test Case 6: Array with negative numbers
    nums6 = [-5, -2, -8, -1, -9, -4, -7, -6, -3]
    expected6 = [-9, -8, -7, -6, -5, -4, -3, -2, -1]
    result6 = solver.sortArray(nums6)
    print(f"Input: {nums6}, Expected: {expected6}, Result: {result6}, {'Pass' if result6 == expected6 else 'Fail'}")

    # Test Case 7: Array with mixed positive and negative numbers
    nums7 = [-1, 5, -3, 2, 0, -2, 4]
    expected7 = [-3, -2, -1, 0, 2, 4, 5]
    result7 = solver.sortArray(nums7)
    print(f"Input: {nums7}, Expected: {expected7}, Result: {result7}, {'Pass' if result7 == expected7 else 'Fail'}")

    # Test Case 8: Array with all same elements
    nums8 = [7, 7, 7, 7, 7]
    expected8 = [7, 7, 7, 7, 7]
    result8 = solver.sortArray(nums8)
    print(f"Input: {nums8}, Expected: {expected8}, Result: {result8}, {'Pass' if result8 == expected8 else 'Fail'}")

    # Test Case 9: Larger array with various numbers
    import random
    random.seed(42)  # For reproducible results
    nums9 = [random.randint(-100, 100) for _ in range(50)]
    expected9 = sorted(nums9)  # Use built-in sorted for expected result
    result9 = solver.sortArray(nums9.copy()) # Sort a copy to compare
    print(f"Input (first 10): {nums9[:10]}..., Expected (first 10): {expected9[:10]}..., Result (first 10): {result9[:10]}..., {'Pass' if result9 == expected9 else 'Fail'}")
````

**Explanation of the Python Conversion:**

- **`function swap(items, l, r)` to `def swap(items, l, r):`**: JavaScript's `function` keyword is replaced with Python's `def` to define a function.
- **`const temp = items[l];` etc. to `temp = items[l]` etc.**: JavaScript's `const` for declaring a constant variable is simply removed in Python. Variable assignment is straightforward.
- **`let pivot = items[end], s = start` to `pivot = items[end]; s = start`**: JavaScript allows declaring multiple variables with `let` in one line. In Python, it's more common to declare them on separate lines for better readability, separated by semicolons if kept on the same line.
- **`for(let i = start; i < end; i++)` to `for i in range(start, end):`**: JavaScript's `for` loop syntax is translated to Python's `for i in range(start, end):`. Note that `range(start, end)` in Python iterates from `start` up to (but not including) `end`, which matches the behavior of the JavaScript loop.
- **`s++` to `s += 1`**: JavaScript's increment operator `++` is replaced with Python's `+=` for incrementing.
- **`const pIdx = partition(items, left, right)` to `pIdx = partition(items, left, right)`**: Similar to the `temp` variable, `const` is removed.
- The overall structure and logic of the `partition` and `quickSort` functions remain the same, as they implement the standard Quick Sort algorithm with the last element as the pivot.
- The `sortArray` function in Python is defined as a method within the `Solution` class, which is a common practice for LeetCode solutions.
