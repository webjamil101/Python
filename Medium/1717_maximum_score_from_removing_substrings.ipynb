1717. Maximum Score From Removing Substrings
      Medium
      You are given a string s and two integers x and y. You can perform two types of operations any number of times.

Remove substring "ab" and gain x points.
For example, when removing "ab" from "cabxbae" it becomes "cxbae".
Remove substring "ba" and gain y points.
For example, when removing "ba" from "cabxbae" it becomes "cabxe".
Return the maximum points you can gain after applying the above operations on s.

Example 1:

Input: s = "cdbcbbaaabab", x = 4, y = 5
Output: 19
Explanation:

- Remove the "ba" underlined in "cdbcbbaaabab". Now, s = "cdbcbbaaab" and 5 points are added to the score.
- Remove the "ab" underlined in "cdbcbbaaab". Now, s = "cdbcbbaa" and 4 points are added to the score.
- Remove the "ba" underlined in "cdbcbbaa". Now, s = "cdbcba" and 5 points are added to the score.
- Remove the "ba" underlined in "cdbcba". Now, s = "cdbc" and 5 points are added to the score.
  Total score = 5 + 4 + 5 + 5 = 19.
  Example 2:

Input: s = "aabbaaxybbaabb", x = 5, y = 4
Output: 20

Constraints:

1 <= s.length <= 105
1 <= x, y <= 104
s consists of lowercase English letters.


```python
class Solution:
    def maximumGain(self, s: str, x: int, y: int) -> int:
        # Helper function to remove one pattern greedily
        def remove_pattern(s, first_char, second_char, score):
            stack = []
            local_score = 0
            for ch in s:
                if stack and stack[-1] == first_char and ch == second_char:
                    stack.pop()
                    local_score += score
                else:
                    stack.append(ch)
            return ''.join(stack), local_score

        total_score = 0
        # Remove the higher score pattern first
        if x > y:
            # Remove "ab" first
            s, gained_x = remove_pattern(s, 'a', 'b', x)
            s, gained_y = remove_pattern(s, 'b', 'a', y)
        else:
            # Remove "ba" first
            s, gained_y = remove_pattern(s, 'b', 'a', y)
            s, gained_x = remove_pattern(s, 'a', 'b', x)

        total_score = gained_x + gained_y
        return total_score
def test():
    sol = Solution()

    assert sol.maximumGain("cdbcbbaaabab", 4, 5) == 19
    assert sol.maximumGain("aabbaaxybbaabb", 5, 4) == 20
    assert sol.maximumGain("ab", 10, 5) == 10  # Only one "ab"
    assert sol.maximumGain("ba", 3, 8) == 8   # Only one "ba"
    assert sol.maximumGain("ababab", 2, 1) == 6  # Remove all "ab"s
    assert sol.maximumGain("baba", 1, 2) == 4    # Remove "ba" then "ba"
    assert sol.maximumGain("", 1, 1) == 0        # Empty string
    assert sol.maximumGain("aaaaa", 10, 10) == 0 # No "ab"/"ba"
    assert sol.maximumGain("bbbbb", 5, 5) == 0   # No valid pair

    print("âœ… All test cases passed!")

test()
```

```python
class Solution:
    def maximumGain(self, s: str, x: int, y: int) -> int:
        total_score = 0

        def remove_and_score(current_s: str, first_char: str, second_char: str, points: int) -> tuple[str, int]:
            nonlocal total_score
            stack = []
            current_gain = 0
            for char in current_s:
                if stack and stack[-1] == first_char and char == second_char:
                    stack.pop()
                    current_gain += points
                else:
                    stack.append(char)
            return "".join(stack), current_gain

        if x >= y:
            # Prioritize removing "ab"
            s_after_ab, gain_ab = remove_and_score(s, 'a', 'b', x)
            total_score += gain_ab

            # Then remove "ba" from the remaining string
            _, gain_ba = remove_and_score(s_after_ab, 'b', 'a', y)
            total_score += gain_ba
        else:
            # Prioritize removing "ba"
            s_after_ba, gain_ba = remove_and_score(s, 'b', 'a', y)
            total_score += gain_ba

            # Then remove "ab" from the remaining string
            _, gain_ab = remove_and_score(s_after_ba, 'a', 'b', x)
            total_score += gain_ab

        return total_score
```
