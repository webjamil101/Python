You are given two 0-indexed arrays `nums1` and `nums2` of length `n`, both of which are **permutations of** `[0, 1, ..., n - 1]`.

A **good triplet** is a set of 3 distinct values which are present in increasing order **by position** in both `nums1` and `nums2`.

In other words, if we define:

- `pos1[v]` as the index of value `v` in `nums1`, and
- `pos2[v]` as the index of value `v` in `nums2`,

Then a **good triplet** is a set `(x, y, z)` such that:

```
0 <= x, y, z <= n - 1,
x, y, z are distinct,
pos1[x] < pos1[y] < pos1[z],
pos2[x] < pos2[y] < pos2[z]
```

Your task is to **return the total number of good triplets**.

---

## Example 1:

**Input:**

```python
nums1 = [2, 0, 1, 3]
nums2 = [0, 1, 2, 3]
```

**Output:**

```
1
```

**Explanation:**  
There are 4 total triplets `(x, y, z)` such that `pos1[x] < pos1[y] < pos1[z]`:

- (2, 0, 1)
- (2, 0, 3)
- (2, 1, 3)
- (0, 1, 3)

Out of those, only the triplet **(0, 1, 3)** satisfies `pos2[x] < pos2[y] < pos2[z]`.

---

## Example 2:

**Input:**

```python
nums1 = [4, 0, 1, 3, 2]
nums2 = [4, 1, 0, 2, 3]
```

**Output:**

```
4
```

**Explanation:**  
The 4 good triplets are:

- (4, 0, 3)
- (4, 0, 2)
- (4, 1, 3)
- (4, 1, 2)

---

## Constraints:

- `n == nums1.length == nums2.length`
- `3 <= n <= 10⁵`
- `0 <= nums1[i], nums2[i] <= n - 1`
- `nums1` and `nums2` are **permutations** of `[0, 1, ..., n - 1]`
