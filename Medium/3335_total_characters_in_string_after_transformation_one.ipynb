# 3335. Total Characters in String After Transformations I

# Medium

You are given a string s and an integer t, representing the number of transformations to perform. In one transformation, every character in s is replaced according to the following rules:

If the character is 'z', replace it with the string "ab".
Otherwise, replace it with the next character in the alphabet. For example, 'a' is replaced with 'b', 'b' is replaced with 'c', and so on.
Return the length of the resulting string after exactly t transformations.

> Since the answer may be very large, return it modulo 109 + 7.

# Example 1:

```
Input: s = "abcyy", t = 2

Output: 7

Explanation:

First Transformation (t = 1):
'a' becomes 'b'
'b' becomes 'c'
'c' becomes 'd'
'y' becomes 'z'
'y' becomes 'z'
String after the first transformation: "bcdzz"
Second Transformation (t = 2):
'b' becomes 'c'
'c' becomes 'd'
'd' becomes 'e'
'z' becomes "ab"
'z' becomes "ab"
String after the second transformation: "cdeabab"
Final Length of the string: The string is "cdeabab", which has 7 characters.
```

# Example 2:

```
Input: s = "azbk", t = 1

Output: 5

Explanation:
First Transformation (t = 1):
'a' becomes 'b'
'z' becomes "ab"
'b' becomes 'c'
'k' becomes 'l'
String after the first transformation: "babcl"
Final Length of the string: The string is "babcl", which has 5 characters.
```

# Constraints:

- 1 <= s.length <= 105
- s consists only of lowercase English letters.
- 1 <= t <= 105


````python
M = int(1e9 + 7)

def length_after_transformations(s: str, t: int) -> int:
    """Returns the transformed length of string s after t iterations."""
    mp = [0] * 26

    for ch in s:
        mp[ord(ch) - ord('a')] += 1

    for _ in range(t):  # O(t)
        temp = [0] * 26

        for i in range(26):  # O(26)
            ch = chr(i + ord('a'))
            freq = mp[i]

            if ch != 'z':
                temp[ord(ch) - ord('a') + 1] = (temp[ord(ch) - ord('a') + 1] + freq) % M
            else:
                temp[0] = (temp[0] + freq) % M
                temp[1] = (temp[1] + freq) % M

        mp = temp

    return sum(mp) % M

# **Test Cases**
test_cases = [
    ("abc", 1),     # Normal case
    ("zz", 2),      # Transforming 'z' case
    ("aaaa", 3),    # Only one character present
    ("xyz", 5),     # Random case
    ("z", 10),      # Single character 'z'
    ("a", 100),     # Single character with large transformations
    ("", 5),        # Empty string edge case
]

for s, t in test_cases:
    print(f"Input: s={s}, t={t} -> Output: {length_after_transformations(s, t)}")
````

````python
class Solution:
    def __init__(self):
        self.M = int(1e9 + 7)

    def lengthAfterTransformations(self, s: str, t: int) -> int:
        """Returns the transformed length of string s after t iterations."""
        mp = [0] * 26

        for ch in s:
            mp[ord(ch) - ord('a')] += 1

        for _ in range(t):  # O(t)
            temp = [0] * 26

            for i in range(26):  # O(26)
                ch = chr(i + ord('a'))
                freq = mp[i]

                if ch != 'z':
                    temp[ord(ch) - ord('a') + 1] = (temp[ord(ch) - ord('a') + 1] + freq) % self.M
                else:
                    temp[0] = (temp[0] + freq) % self.M
                    temp[1] = (temp[1] + freq) % self.M

            mp = temp

        return sum(mp) % self.M

# **Testing Object-Oriented Version**
solution = Solution()

test_cases = [
    ("abc", 1), 
    ("zz", 2), 
    ("aaaa", 3), 
    ("xyz", 5), 
    ("z", 10), 
    ("a", 100), 
    ("", 5),    
]

for s, t in test_cases:
    print(f"Input: s={s}, t={t} -> Output: {solution.lengthAfterTransformations(s, t)}")
````
