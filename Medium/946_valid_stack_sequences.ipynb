# 946. Validate Stack Sequences

**Medium**

Given two integer arrays pushed and popped each with distinct values, return true if this could have been the result of a sequence of push and pop operations on an initially empty stack, or false otherwise.

# Example 1:

```
Input: pushed = [1,2,3,4,5], popped = [4,5,3,2,1]
Output: true
```

Explanation: We might do the following sequence:
push(1), push(2), push(3), push(4),
pop() -> 4,
push(5),
pop() -> 5, pop() -> 3, pop() -> 2, pop() -> 1

# Example 2:

```Python
Input: pushed = [1,2,3,4,5], popped = [4,3,5,1,2]
Output: false
```

**Explanation:** 1 cannot be popped before 2.

**Constraints:**

- 1 <= pushed.length <= 1000
- 0 <= pushed[i] <= 1000
- All the elements of pushed are unique.
- popped.length == pushed.length
- popped is a permutation of pushed.


````python
class SolutionStackSimulation:
    def validateStackSequences(self, pushed: list[int], popped: list[int]) -> bool:
        simulated_stack = []
        j = 0  # Pointer for the popped array

        for x in pushed:
            simulated_stack.append(x)
            while simulated_stack and simulated_stack[-1] == popped[j]:
                simulated_stack.pop()
                j += 1

        return not simulated_stack
		
def test_validate_sequences():
    s = SolutionStackSimulation()

    assert s.validateStackSequences([1,2,3,4,5], [4,5,3,2,1]) == True
    assert s.validateStackSequences([1,2,3,4,5], [4,3,5,1,2]) == False
    assert s.validateStackSequences([], []) == True
    assert s.validateStackSequences([1], [1]) == True
    assert s.validateStackSequences([1], [2]) == False
    assert s.validateStackSequences([2,1,0], [1,2,0]) == True
    assert s.validateStackSequences([1,2,3], [1,3,2]) == True
    assert s.validateStackSequences([1,2,3], [2,3,1]) == False

    print("All tests passed.")

test_validate_sequences()
````

````python
import collections

class Solution:
    def validateStackSequences(self, pushed: list[int], popped: list[int]) -> bool:
        """
        Validates if a popped sequence is possible for a given pushed sequence.
        The approach is to simulate the stack operations.
        
        Args:
            pushed: A list of integers representing the push order.
            popped: A list of integers representing the proposed pop order.

        Returns:
            True if the popped sequence is valid, False otherwise.
        """
        stack = []
        j = 0  # Pointer for the popped sequence
        
        for num in pushed:
            stack.append(num)
            # While the stack is not empty and the top of the stack
            # matches the current element in the popped sequence, pop it.
            while stack and j < len(popped) and stack[-1] == popped[j]:
                stack.pop()
                j += 1
        
        # If the stack is empty, it means all elements were popped in the
        # correct order, so the sequence is valid.
        return not stack
````
