# 1007. Minimum Domino Rotations For Equal Row

# Medium

In a row of dominoes, tops[i] and bottoms[i] represent the top and bottom halves of the ith domino. (A domino is a tile with two numbers from 1 to 6 - one on each half of the tile.)

We may rotate the ith domino, so that tops[i] and bottoms[i] swap values.

Return the minimum number of rotations so that all the values in tops are the same, or all the values in bottoms are the same.

If it cannot be done, return -1.

# Example 1:

```nput: tops = [2,1,2,4,2,2], bottoms = [5,2,6,2,3,2]
Output: 2
Explanation:
The first figure represents the dominoes as given by tops and bottoms: before we do any rotations.
If we rotate the second and fourth dominoes, we can make every value in the top row equal to 2, as indicated by the second figure.

```

# Example 2:

```
Input: tops = [3,5,1,2,3], bottoms = [3,6,3,3,4]
Output: -1
Explanation:
In this case, it is not possible to rotate the dominoes to make one row of values equal.
```

# Constraints:

- 2 <= tops.length <= 2 \* 104
- bottoms.length == tops.length
- 1 <= tops[i], bottoms[i] <= 6


There are multiple ways to solve the **Minimum Domino Rotations** problem efficiently. Below are the different approaches:

---

### **1. Brute Force Approach (`O(n^2)`)**

#### **Idea**

- Try making all values in the `tops` array equal to `1-6` and count rotations.
- Try making all values in the `bottoms` array equal to `1-6` and count rotations.
- Return the minimum rotations needed.

#### **Time Complexity:** `O(n^2)` (not optimal for large inputs)

#### **Implementation**

```python
def minDominoRotationsBruteForce(tops, bottoms):
    n = len(tops)
    min_rotations = float('inf')

    for target in range(1, 7):  # Try numbers from 1 to 6
        top_rotations = bottom_rotations = 0
        possible = True

        for i in range(n):
            if tops[i] != target and bottoms[i] != target:
                possible = False
                break
            elif tops[i] != target:
                top_rotations += 1
            elif bottoms[i] != target:
                bottom_rotations += 1

        if possible:
            min_rotations = min(min_rotations, top_rotations, bottom_rotations)

    return min_rotations if min_rotations != float('inf') else -1

# Example Usage
print(minDominoRotationsBruteForce([2,1,2,4,2,2], [5,2,6,2,3,2]))  # Output: 2
print(minDominoRotationsBruteForce([3,5,1,2,3], [3,6,3,3,4]))  # Output: -1
```

---

### **2. Optimized Approach (`O(n)`)**

#### **Idea**

- Focus only on two numbers: `tops[0]` or `bottoms[0]`.
- Try making all values equal to `tops[0]`, then try `bottoms[0]`.
- Count rotations needed and return the minimum.

#### **Time Complexity:** `O(n)`

#### **Implementation**

```python
def minDominoRotationsOptimized(tops, bottoms):
    def check(target):
        top_rotations = bottom_rotations = 0

        for i in range(len(tops)):
            if tops[i] != target and bottoms[i] != target:
                return float('inf')  # Impossible to make all elements equal
            elif tops[i] != target:
                top_rotations += 1
            elif bottoms[i] != target:
                bottom_rotations += 1

        return min(top_rotations, bottom_rotations)

    rotations = min(check(tops[0]), check(bottoms[0]))
    return rotations if rotations != float('inf') else -1

# Example Usage
print(minDominoRotationsOptimized([2,1,2,4,2,2], [5,2,6,2,3,2]))  # Output: 2
print(minDominoRotationsOptimized([3,5,1,2,3], [3,6,3,3,4]))  # Output: -1
```

---

### **3. Comparing Both Approaches**

| Approach               | Time Complexity | Space Complexity | Efficiency                       |
| ---------------------- | --------------- | ---------------- | -------------------------------- |
| **Brute Force**        | `O(n^2)`        | `O(1)`           | **Too slow for large inputs**    |
| **Optimized (`O(n)`)** | `O(n)`          | `O(1)`           | **Best choice for large inputs** |

The **optimized approach (`O(n)`)** is the best because it efficiently determines if rotating dominoes can achieve a uniform row.


````python
from typing import List

class Solution:
    def minDominoRotations(self, tops: List[int], bottoms: List[int]) -> int:
        def check(target: int) -> int:
            """
            Checks the number of rotations required to make all values in tops or bottoms equal to target.
            If it's not possible, returns infinity.
            """
            top_rotations = bottom_rotations = 0

            for i in range(len(tops)):
                if tops[i] != target and bottoms[i] != target:
                    return float('inf')  # Impossible to make all elements equal
                elif tops[i] != target:
                    top_rotations += 1
                elif bottoms[i] != target:
                    bottom_rotations += 1

            return min(top_rotations, bottom_rotations)

        # Try to make all elements equal to tops[0] or bottoms[0]
        rotations = min(check(tops[0]), check(bottoms[0]))
        return rotations if rotations != float('inf') else -1


# **Edge Cases & Test Cases**
def run_tests():
    test_cases = [
        ([2, 1, 2, 4, 2, 2], [5, 2, 6, 2, 3, 2], 2),  # Normal case
        ([3, 5, 1, 2, 3], [3, 6, 3, 3, 4], -1),  # No solution
        ([1, 1, 1, 1], [1, 1, 1, 1], 0),  # Already uniform
        ([1, 2, 1, 2, 1], [2, 1, 2, 1, 2], 1),  # Alternating dominoes
        ([1, 1, 1, 2], [2, 2, 2, 1], 1),  # Minimal swap required
        ([6, 6, 6, 6], [6, 6, 6, 6], 0),  # All elements are already the same
        ([1, 2, 3, 4], [4, 3, 2, 1], -1),  # Impossible to match any number
        ([5, 2, 2, 2, 5], [2, 5, 5, 5, 2], 2),  # Multiple possible solutions
    ]

    for tops, bottoms, expected in test_cases:
        solver = Solution()
        result = solver.minDominoRotations(tops, bottoms)
        print(f"Input Tops: {tops}, Bottoms: {bottoms} | Expected: {expected}, Got: {result} | {'Pass' if result == expected else 'Fail'}")

# Run the test cases
run_tests()
````

````python
from typing import List

class MinDominoRotationsBruteForce:
    def __init__(self, tops: List[int], bottoms: List[int]):
        self.tops = tops
        self.bottoms = bottoms
        self.n = len(tops)

    def find_min_rotations(self) -> int:
        min_rotations = float('inf')

        # Try numbers from 1 to 6 as the target
        for target in range(1, 7):
            top_rotations = bottom_rotations = 0
            possible = True

            for i in range(self.n):
                if self.tops[i] != target and self.bottoms[i] != target:
                    possible = False
                    break
                elif self.tops[i] != target:
                    top_rotations += 1
                elif self.bottoms[i] != target:
                    bottom_rotations += 1

            if possible:
                min_rotations = min(min_rotations, top_rotations, bottom_rotations)

        return min_rotations if min_rotations != float('inf') else -1


# **Edge Cases & Test Cases**
def run_tests():
    test_cases = [
        ([2, 1, 2, 4, 2, 2], [5, 2, 6, 2, 3, 2], 2),  # Normal case
        ([3, 5, 1, 2, 3], [3, 6, 3, 3, 4], -1),  # No solution possible
        ([1, 1, 1, 1], [1, 1, 1, 1], 0),  # Already uniform, no rotations needed
        ([1, 2, 1, 2, 1], [2, 1, 2, 1, 2], 1),  # Alternating pattern
        ([6, 6, 6, 6], [6, 6, 6, 6], 0),  # All elements are the same
        ([1, 2, 3, 4], [4, 3, 2, 1], -1),  # No single common value possible
        ([5, 2, 2, 2, 5], [2, 5, 5, 5, 2], 2),  # Multiple valid swaps available
    ]

    for tops, bottoms, expected in test_cases:
        solver = MinDominoRotationsBruteForce(tops, bottoms)
        result = solver.find_min_rotations()
        print(f"Input Tops: {tops}, Bottoms: {bottoms} | Expected: {expected}, Got: {result} | {'Pass' if result == expected else 'Fail'}")

# Run the test cases
run_tests()
````
