**1493. Longest Subarray of 1's After Deleting One Element**

**Medium**

**Companies**:Yandex, Google, Meta

Given a binary array nums, you should delete one element from it.

Return the size of the longest non-empty subarray containing only 1's in the resulting array. Return 0 if there is no such subarray.

**Example 1:**

```python
Input: nums = [1,1,0,1]
Output: 3
```

**Explanation:** After deleting the number in position 2, [1,1,1] contains 3 numbers with value of 1's.

**Example 2:**

```python
Input: nums = [0,1,1,1,0,1,1,0,1]
Output: 5
```

**Explanation:** After deleting the number in position 4, [0,1,1,1,1,1,0,1] longest subarray with value of 1's is [1,1,1,1,1].
**Example 3:**

```python
Input: nums = [1,1,1]
Output: 2
```

**Explanation:** You must delete one element.

**Constraints:**

- 1 <= nums.length <= 105
- nums[i] is either 0 or 1.


````python
class Solution:
    def longestSubarray(self, nums: list[int]) -> int:
        n = len(nums)
        result = 0
        count_zero = 0

        # Helper function to find max length after skipping one element
        def find_max(arr, skip_idx):
            current_length = 0
            max_length = 0
            for i in range(len(arr)):
                if i == skip_idx:
                    continue
                if arr[i] == 1:
                    current_length += 1
                    max_length = max(max_length, current_length)
                else:
                    current_length = 0
            return max_length

        for i in range(n):
            if nums[i] == 0:
                count_zero += 1
                result = max(result, find_max(nums, i))

        if count_zero == 0:
            return n - 1
        
        return result
````

````python
class Solution:
    def longestSubarray(self, nums: List[int]) -> int:
        n = len(nums)
        
        # Handle case with no zeros
        if 0 not in nums:
            return n - 1
        
        # Create prefix array (consecutive 1's from left)
        prefix = [0] * n
        prefix[0] = nums[0]
        for i in range(1, n):
            prefix[i] = prefix[i-1] + 1 if nums[i] == 1 else 0
        
        # Create suffix array (consecutive 1's from right)
        suffix = [0] * n
        suffix[n-1] = nums[n-1]
        for i in range(n-2, -1, -1):
            suffix[i] = suffix[i+1] + 1 if nums[i] == 1 else 0
        
        result = 0
        for i in range(n):
            if nums[i] == 0:
                left = prefix[i-1] if i > 0 else 0
                right = suffix[i+1] if i < n-1 else 0
                result = max(result, left + right)
        
        return result
````
