**_2256. Minimum Average Difference_**

**_Medium_**

**_Companies_**

You are given a 0-indexed integer array nums of length n.

The average difference of the index i is the absolute difference between the average of the first i + 1 elements of nums and the average of the last n - i - 1 elements. Both averages should be rounded down to the nearest integer.

Return the index with the minimum average difference. If there are multiple such indices, return the smallest one.

> **_Note:_**

The absolute difference of two numbers is the absolute value of their difference.
The average of n elements is the sum of the n elements divided (integer division) by n.
The average of 0 elements is considered to be 0.

**_Example 1:_**

```python
Input: nums = [2,5,3,9,5,3]
Output: 3
```

**_Explanation:_**

- The average difference of index 0 is: |2 / 1 - (5 + 3 + 9 + 5 + 3) / 5| = |2 / 1 - 25 / 5| = |2 - 5| = 3.
- The average difference of index 1 is: |(2 + 5) / 2 - (3 + 9 + 5 + 3) / 4| = |7 / 2 - 20 / 4| = |3 - 5| = 2.
- The average difference of index 2 is: |(2 + 5 + 3) / 3 - (9 + 5 + 3) / 3| = |10 / 3 - 17 / 3| = |3 - 5| = 2.
- The average difference of index 3 is: |(2 + 5 + 3 + 9) / 4 - (5 + 3) / 2| = |19 / 4 - 8 / 2| = |4 - 4| = 0.
- The average difference of index 4 is: |(2 + 5 + 3 + 9 + 5) / 5 - 3 / 1| = |24 / 5 \*\*\*- 3 / 1| = |4 - 3| = 1.
- The average difference of index 5 is: |(2 + 5 + 3 + 9 + 5 + 3) / 6 - 0| = |27 / 6 - 0| = |4 - 0| = 4.
  The average difference of index 3 is the minimum average difference so return 3.

**Example 2:**

```python
Input: nums = [0]
Output: 0
```

**_Explanation:_**
The only index is 0 so return 0.
The average difference of index 0 is: |0 / 1 - 0| = |0 - 0| = 0.

**_Constraints:_**

- 1 <= nums.length <= 105
- 0 <= nums[i] <= 105


````python
class Solution:
    def minimumAverageDifference(self, nums: List[int]) -> int:
        # Initialize prefix sum, suffix sum, and the number of elements
        prefix_sum, suffix_sum = 0, sum(nums)
        num_elements = len(nums)
      
        # Initialize variables to track the minimum average difference and its index
        min_index, min_difference = 0, float('inf')
      
        # Iterate through the list to calculate the differences
        for index, value in enumerate(nums):
            # Update the prefix and suffix sums
            prefix_sum += value
            suffix_sum -= value
          
            # Calculate the average of the prefix and suffix
            prefix_avg = prefix_sum // (index + 1)
            # Calculate the average of suffix; handle the case when it becomes empty
            suffix_avg = 0 if num_elements - index - 1 == 0 else suffix_sum // (num_elements - index - 1)
          
            # Calculate the absolute difference of the averages
            current_difference = abs(prefix_avg - suffix_avg)
          
            # If the current difference is smaller than the recorded minimum, update the result
            if current_difference < min_difference:
                min_index = index
                min_difference = current_difference
      
        # Return the index of the element that gives the minimum average difference
        return min_index
````

````python
import math

class Solution:
    def minimumAverageDifference(self, nums: List[int]) -> int:
        n = len(nums)
        min_avg_diff = float('inf')
        result_index = 0

        for i in range(n):
            # Calculate prefix average
            prefix_sum = sum(nums[:i + 1])
            prefix_count = i + 1
            prefix_avg = prefix_sum // prefix_count

            # Calculate suffix average
            suffix_sum = sum(nums[i + 1:])
            suffix_count = n - (i + 1)
            suffix_avg = 0
            if suffix_count > 0:
                suffix_avg = suffix_sum // suffix_count
            
            # Calculate average difference
            current_avg_diff = abs(prefix_avg - suffix_avg)

            # Update minimum average difference and result index
            if current_avg_diff < min_avg_diff:
                min_avg_diff = current_avg_diff
                result_index = i
        
        return result_index
````
